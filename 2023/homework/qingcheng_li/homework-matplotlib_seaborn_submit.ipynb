## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
#计算每年乘客数
year_sum= data[['year', 'passengers']].groupby('year').sum().reset_index()
year_sum
```

```python
plt.figure
plt.plot(year_sum['year'], year_sum['passengers'],linestyle='-')
plt.xlabel('year')
plt.ylabel('number')
plt.xticks(year_sum['year'])
plt.title('the number of the passengers each year')
```

```python
month_sum= data[['month', 'passengers']].groupby('month').sum().reset_index()
month_sum
```

```python
plt.bar(month_sum['month'],month_sum['passengers'])
plt.xlabel('month')
plt.ylabel('number')
plt.xticks(month_sum['month'],range(1,13))
plt.title('the number of the passengers each month')
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
#计算每个pclass多少人
pclass_counts=data['pclass'].value_counts()
pclass_counts
```

```python
#计算每个pclass活下来多少人
data1=data[['survived', 'pclass']].groupby(['pclass']).sum()
data1
```

```python
survived_prop=data1['survived']/pclass_counts
survived_prop
unsurvived_prop=1-survived_prop
unsurvived_prop
```

```python
sns.barplot(x=[1,2,3],y=survived_prop,label='survived_prop')
sns.barplot(x=[1,2,3],y=1-survived_prop,alpha=0.5,bottom=survived_prop,label='unsurvived_prop')
plt.xlabel('pclass')
plt.ylabel('prop')
plt.legend()
plt.title('prop of the survived in different pclass')
```

```python
sex_counts=data['sex'].value_counts()
data1=data[['survived', 'sex']].groupby(['sex']).sum()
survived_prop=data1['survived']/sex_counts
unsurvived_prop=1-survived_prop
survived_prop
```

```python
sns.barplot(x=[1,2],y=survived_prop,label='survived_prop')
sns.barplot(x=[1,2],y=1-survived_prop,alpha=0.5,bottom=survived_prop,label='unsurvived_prop')
plt.xticks([0,1],['female','male'])
plt.xlabel('sex')
plt.ylabel('prop')
plt.legend()
plt.title('prop of the survived in different pclass')
```
