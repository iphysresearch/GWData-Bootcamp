## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
#提示
```

```python
#我的代码
plt.figure(figsize=(13,9))

passengers_year = data.groupby('year')['passengers'].sum()
plt.subplot(2,2,1)
plt.title("The number of passengers in each year")   # 标题
plt.xlabel("year")       # 横坐标
plt.ylabel("number")     # 纵坐标
plt.plot(passengers_year.index, passengers_year.values) 

plt.subplot(2,2,2)
passengers_month = data.groupby('month')['passengers'].sum()
plt.title("The number of passengers in each month")   
plt.xlabel("month")       
plt.ylabel("number")       
plt.bar(passengers_month.index, passengers_month.values) 
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
#提示
```

```python
#我的代码1.萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
plt.figure(figsize=(7,7))

plt.subplot(2,2,1)
plt.title("sepal_length & petal_length")   
plt.xlabel("sepal_length")       
plt.ylabel("petal_length")       
plt.scatter(data.sepal_length, data.petal_length,s=6) 

plt.subplot(2,2,2)
plt.title("sepal_length & petal_width")   
plt.xlabel("sepal_length")       
plt.ylabel("petal_width")       
plt.scatter(data.sepal_length, data.petal_width,s=6) 

plt.subplot(2,2,3)
plt.title("sepal_width & petal_length")   
plt.xlabel("sepal_width")       
plt.ylabel("petal_length")       
plt.scatter(data.sepal_width, data.petal_length,s=6) 

plt.subplot(2,2,4)
plt.title("sepal_width & petal_width")   
plt.xlabel("sepal_width")       
plt.ylabel("petal_width")       
plt.scatter(data.sepal_width, data.petal_width,s=6) 

plt.tight_layout()
```

```python
#提示
```

```python
#我的代码2.不同种类（species）鸢尾花萼片和花瓣的大小关系
color_list = {'setosa': 'lightblue', 'versicolor':'red', 'virginica':'black'}
colors = data['species'].map(color_list)

plt.figure(figsize=(7,7))

plt.subplot(2,2,1)
plt.title("sepal_length & petal_length")   
plt.xlabel("sepal_length")       
plt.ylabel("petal_length")       
plt.scatter(data.sepal_length, data.petal_length,s=4,color=colors) 

plt.subplot(2,2,2)
plt.title("sepal_length & petal_width")   
plt.xlabel("sepal_length")       
plt.ylabel("petal_width")       
plt.scatter(data.sepal_length, data.petal_width,s=4,color=colors) 

plt.subplot(2,2,3)
plt.title("sepal_width & petal_length")   
plt.xlabel("sepal_width")       
plt.ylabel("petal_length")       
plt.scatter(data.sepal_width, data.petal_length,s=4,color=colors) 

plt.subplot(2,2,4)
plt.title("sepal_width & petal_width")   
plt.xlabel("sepal_width")       
plt.ylabel("petal_width")       
plt.scatter(data.sepal_width, data.petal_width,s=4,color=colors) 

plt.tight_layout()
```

```python
#提示
```

```python
#我的代码3：不同种类箱图
plt.figure(figsize=(4.5,10))
plt.subplot(3,1,1)
plt.title("The box with 4 features for the setosa flowers")   
plt.xlabel("features of the setosa flowers")       
plt.ylabel("Values")       
plt.boxplot([data[data['species'] == 'setosa']['sepal_length'],
            data[data['species'] == 'setosa']['sepal_width'],
            data[data['species'] == 'setosa']['petal_length'],
            data[data['species'] == 'setosa']['petal_width']],
            boxprops={'color':'blue'},patch_artist=True,medianprops = {'color': 'yellow'},
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

plt.subplot(3,1,2)
plt.title("The box with 4 features for the versicolor flowers")   
plt.xlabel("features of the versicolor flowers")       
plt.ylabel("Values")       
plt.boxplot([data[data['species'] == 'versicolor']['sepal_length'],
            data[data['species'] == 'versicolor']['sepal_width'],
            data[data['species'] == 'versicolor']['petal_length'],
            data[data['species'] == 'versicolor']['petal_width']],
            boxprops={'color':'blue'},patch_artist=True,medianprops = {'color': 'yellow'},
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

plt.subplot(3,1,3)
plt.title("The box with 4 features for the virginica flowers")   
plt.xlabel("features of the virginica flowers")       
plt.ylabel("Values")       
plt.boxplot([data[data['species'] == 'virginica']['sepal_length'],
            data[data['species'] == 'virginica']['sepal_width'],
            data[data['species'] == 'virginica']['petal_length'],
            data[data['species'] == 'virginica']['petal_width']],
            boxprops={'color':'blue'},patch_artist=True,medianprops = {'color': 'yellow'},
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.tight_layout()
```

## 练习3：餐厅小费情况分析 (7个题)

1. 小费和总消费之间的关系（提示：散点图+回归分析）
2. 男性顾客和女性顾客，谁更慷慨（提示：箱图或者提琴图）
3. 抽烟与否是否会对小费金额产生影响（提示：箱图或者提琴图）
4. 工作日和周末，什么时候顾客给的小费更慷慨（提示：箱图或者提琴图）
5. 午饭和晚饭，哪一顿顾客更愿意给小费（提示：箱图或者提琴图）
6. 就餐人数是否会对慷慨度产生影响（提示：箱图或者提琴图）
7. 性别+抽烟的组合因素对慷慨度的影响（提示：统计柱状图）

```python
data = sns.load_dataset("tips")
data.head()
# 总消费，小费，性别，吸烟与否，就餐星期，就餐时间，就餐人数
```

```python
#提示1
```

```python
#我的代码1：小费和总消费之间的关系（提示：散点图+回归分析）
slope, intercept, r_value, p_value, std_err = stats.linregress(data['total_bill'].values, data['tip'].values)
plt.plot(data['total_bill'].values, slope * data['total_bill'].values + intercept, c = 'red')
plt.title("total bill & tip")   
plt.xlabel("total bill")       
plt.ylabel("tip")       
plt.scatter(data.total_bill,data.tip,s=6) 
```

```python
#提示2
```

```python
#我的代码2：男性顾客和女性顾客，谁更慷慨（提示：箱图或者提琴图）
#箱图
plt.title("The box with 2 features for the tip")   
plt.xlabel("Features of the sex")       
plt.ylabel("Values of tip")       
plt.boxplot([data[data['sex'] == 'Female']['tip'],
            data[data['sex'] == 'Male']['tip']],
            boxprops={'color':'blue'},patch_artist=True,medianprops = {'color': 'yellow'},
            labels=['Female','Male'])
```

```python
#提琴图
plt.title("The box with 2 features for the tip")   
plt.xlabel("Features of the sex")       
plt.ylabel("Values of tip")   
plt.violinplot([data[data['sex'] == 'Female']['tip'],
              data[data['sex'] == 'Male']['tip']])
plt.xticks(ticks=[1,2], labels=['Female', 'Male'])
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
#提示1
```

```python
#我的代码1：不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
data['unsurvived'] = 1-data['survived']
data1 = data.groupby('pclass')[['survived', 'unsurvived','pclass']].mean()
print(data1)
data1[['survived','unsurvived']].plot(kind='bar',stacked=True,color=['#5F9EA0','#6A5ACD'])
plt.title("Proportion of survived/unsurvived by pclass(1,2,3)")   
plt.xlabel("pclass")       
plt.ylabel("Proportion of survived/unsurvived")  
plt.xticks(rotation=0)
```

```python
#箱图
plt.boxplot([data1[data1['pclass'] == 1]['survived'],
            data1[data1['pclass'] == 2]['survived'],
            data1[data1['pclass'] == 3]['survived']],
            boxprops={'color':'blue'},patch_artist=True,medianprops = {'color': 'red'},
            labels=['1', '2', '3'])
plt.title("Proportion of survived/unsurvived by pclass(1,2,3)")   
plt.xlabel("pclass")       
plt.ylabel("Proportion of survived/unsurvived") 
```

```python
#提示2
```

```python
#我的代码2：不同性别的幸存比例（提示：箱图或者提琴图）
data2 = data.groupby('sex')[['survived', 'unsurvived']].sum()
data2['total']=data2['survived']+data2['unsurvived']
data22 = data.groupby('sex')[['survived', 'unsurvived']].mean()
data2['survived_prop']=data22['survived']
data2['unsurvived_prop']=data22['unsurvived']
print(data2)

data2[['survived_prop','unsurvived_prop']].plot(kind='bar',stacked=True,color=['#5F9EA0','#6A5ACD'])
plt.title("Proportion of survived/unsurvived by pclass(1,2,3)")   
plt.xlabel("sex")       
plt.ylabel("Proportion of survived/unsurvived")  
plt.xticks(rotation=0)
```
