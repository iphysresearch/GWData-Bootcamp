## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
# 计算年度乘客总量
yearly_passengers = data.groupby('year')['passengers'].sum()

# 绘制折线图
plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_passengers.index, y=yearly_passengers.values)
plt.title('Yearly Total Passengers')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.show()
data = sns.load_dataset("flights")

# 绘制柱状图（去除误差线）
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='passengers', data=data, ci=None)
plt.title('Monthly Passengers Distribution')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset("iris")

import matplotlib.pyplot as plt
import seaborn as sns

# 加载数据集
data = sns.load_dataset("iris")

# 萼片长度和花瓣长度的散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='petal_length', data=data, hue='species')
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()

# 萼片宽度和花瓣长度的散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_width', y='petal_length', data=data, hue='species')
plt.title('Sepal Width vs. Petal Length')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.show()

# 萼片长度和花瓣宽度的散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='petal_width', data=data, hue='species')
plt.title('Sepal Length vs. Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.show()

# 萼片宽度和花瓣宽度的散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_width', y='petal_width', data=data, hue='species')
plt.title('Sepal Width vs. Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.show()
# 绘制箱图或提琴图
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=data)
sns.boxplot(x='species', y='petal_length', data=data, color='lightblue')
plt.title('Sepal and Petal Sizes by Species')
plt.xlabel('Species')
plt.ylabel('Size')
plt.show()
# 绘制六角箱图或核密度估计图
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=data)
sns.violinplot(x='species', y='petal_length', data=data, color='lightblue')
plt.title('Sepal and Petal Size Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Size')
plt.show()
```

## 练习3：餐厅小费情况分析 (7个题)

1. 小费和总消费之间的关系（提示：散点图+回归分析）
2. 男性顾客和女性顾客，谁更慷慨（提示：箱图或者提琴图）
3. 抽烟与否是否会对小费金额产生影响（提示：箱图或者提琴图）
4. 工作日和周末，什么时候顾客给的小费更慷慨（提示：箱图或者提琴图）
5. 午饭和晚饭，哪一顿顾客更愿意给小费（提示：箱图或者提琴图）
6. 就餐人数是否会对慷慨度产生影响（提示：箱图或者提琴图）
7. 性别+抽烟的组合因素对慷慨度的影响（提示：统计柱状图）

```python
data = sns.load_dataset("tips")
data.head()
# 总消费，小费，性别，吸烟与否，就餐星期，就餐时间，就餐人数
```

```python
# 绘制散点图和回归线
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=data)
plt.title('Relationship between Total Bill and Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
# 绘制箱图或提琴图
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=data)
plt.title('Tip by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()
# 绘制箱图或提琴图
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='tip', data=data)
plt.title('Tip by Smoking Status')
plt.xlabel('Smoker')
plt.ylabel('Tip')
plt.show()
# 添加工作日和周末标签
data['is_weekend'] = data['day'].isin(['Sat', 'Sun'])
# 绘制箱图或提琴图
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_weekend', y='tip', data=data)
plt.title('Tip by Weekday/Weekend')
plt.xlabel('Weekday/Weekend')
plt.ylabel('Tip')
plt.xticks([0, 1], ['Weekday', 'Weekend'])
plt.show()
# 添加午饭和晚饭标签
data['meal_time'] = data['time'].map({'Lunch': 'Lunch', 'Dinner': 'Dinner'})
# 绘制箱图或提琴图
plt.figure(figsize=(8, 6))
sns.boxplot(x='meal_time', y='tip', data=data)
plt.title('Tip by Meal Time')
plt.xlabel('Meal Time')
plt.ylabel('Tip')
plt.show()
# 绘制箱图或提琴图
plt.figure(figsize=(8, 6))
sns.boxplot(x='size', y='tip', data=data)
plt.title('Tip by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Tip')
plt.show()
# 绘制统计柱状图
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='tip', hue='smoker', data=data)
plt.title('Tip by Gender and Smoking Status')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()





```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
# 不同仓位等级中幸存和遇难乘客比例的分析（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='survived', data=data)
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()

# 不同性别的幸存比例的分析（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='survived', data=data)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# 幸存和遇难乘客的票价分布分析（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='fare', data=data)
plt.title('Fare Distribution of Survived and Deceased Passengers')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks([0, 1], ['Deceased', 'Survived'])
plt.show()
# 幸存和遇难乘客的年龄分布（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='age', data=data)
plt.title('Age Distribution of Survived and Deceased Passengers')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.xticks([0, 1], ['Deceased', 'Survived'])
plt.show()

# 不同上船港口的乘客仓位等级分布（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='embark_town', y='class', data=data)
plt.title('Passenger Class Distribution by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Class')
plt.show()

# 幸存和遇难乘客堂兄弟姐妹的数量分布（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='sibsp', data=data)
plt.title('Sibling/Spouse Distribution of Survived and Deceased Passengers')
plt.xlabel('Survived')
plt.ylabel('Siblings/Spouses')
plt.xticks([0, 1], ['Deceased', 'Survived'])
plt.show()
# 幸存和遇难乘客父母子女的数量分布（箱图或提琴图）
plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='parch', data=data)
plt.title('Parent/Child Distribution of Survived and Deceased Passengers')
plt.xlabel('Survived')
plt.ylabel('Parent/Child')
plt.xticks([0, 1], ['Deceased', 'Survived'])
plt.show()

# 单独乘船与否和幸存之间的联系（统计柱状图）
plt.figure(figsize=(8, 6))
survived_counts = data[data['survived'] == 1]['alone'].value_counts()
deceased_counts = data[data['survived'] == 0]['alone'].value_counts()
labels = ['Alone', 'Not Alone']
x = range(len(labels))
plt.bar(x, survived_counts, width=0.4, align='center', label='Survived')
plt.bar(x, deceased_counts, width=0.4, align='edge', label='Deceased')
plt.xlabel('Alone')
plt.ylabel('Count')
plt.title('Survival by Alone Status')
plt.xticks(x, labels)
plt.legend()
plt.show()
```
