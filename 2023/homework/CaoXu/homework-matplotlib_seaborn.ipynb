## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
yearly_passengers = data.groupby('year')['passengers'].sum()
plt.plot(yearly_passengers.index, yearly_passengers.values)
plt.xlabel('year')
plt.ylabel('number')
plt.title('The number of passengers in each year')
plt.show()
```

```python
monthly_passengers = data.groupby('month')['passengers'].sum()
plt.bar(monthly_passengers.index, monthly_passengers.values)
plt.xlabel('month')
plt.ylabel('number')
plt.title('The number of passengers in each month')
plt.show()
```

```python
#7. 性别+抽烟的组合因素对慷慨度的影响（提示：统计柱状图）
import numpy as  np
def groupedbarplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 设置每一组柱状图的宽度
    total_width = 0.8
    # 设置每一个柱状图的宽度
    ind_width = total_width / len(y_data_list)

    # 计算每一个柱状图的中心偏移
    alteration = np.arange(-total_width/2+ind_width/2, total_width/2+ind_width/2, ind_width)

    # 分别绘制每一个柱状图
    for i in range(0, len(y_data_list)):
        # 横向散开绘制
        ax.bar(np.arange(len(x_data)) + alteration[i], y_data_list[i], color = colors[i], label = y_data_names[i], width = ind_width)
    ax.set_xticks(np.arange(len(x_data)))
    ax.set_xticklabels(x_data)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right')

#数据分类
grouped_data1 = data[['sex', 'smoker','tip']].groupby(['sex','smoker'])['tip'].mean()

#整理一下
grouped_data = grouped_data1.unstack()
print(grouped_data)

groupedbarplot(x_data = grouped_data.index.values
               , y_data_list = [grouped_data['Yes'].values, grouped_data['No'].values]
               , y_data_names = ['Yes', 'No']
               , colors = ['#539caf', '#7663b0']
               , x_label = 'Sex'
               , y_label = 'Value of tip'
               , title = 'Values by Sex (Male or Female) and Smoker (Yes or No)')
```

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
import pandas as pd
#1.不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
def stackedbarplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 循环绘制堆积柱状图
    for i in range(0, len(y_data_list)):
        if i == 0:
            ax.bar(x_data, y_data_list[i], color = colors[i], align = 'center', label = y_data_names[i])
        else:
            # 采用堆积的方式，除了第一个分类，后面的分类都从前一个分类的柱状图接着画
            # 用归一化保证最终累积结果为1，下面bottom参数表示纵向从哪里开始画
            ax.bar(x_data, y_data_list[i], color = colors[i], bottom = y_data_list[i - 1], align = 'center', label = y_data_names[i])
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right') # 设定图例位置

# 按照pclass分类，计算survived中0和1出现的次数
pclass_data = data.groupby('pclass')['survived'].value_counts().unstack()
#计算比例
pclass_data['total']=pclass_data[1]+pclass_data[0]
pclass_data['survived_prop']=pclass_data[1]/pclass_data['total']
pclass_data['unsurvived_prop']=pclass_data[0]/pclass_data['total']
pclass_data.rename(columns={0: 'unservived', 1: 'survived'}, inplace=True)
print(pclass_data)

stackedbarplot(x_data = pclass_data.index.values
               , y_data_list = [pclass_data['survived_prop'], pclass_data['unsurvived_prop']]
               , y_data_names = ['survived', 'unservived']
               , colors = ['#539caf', '#7663b0']
               , x_label = 'Pclass'
               , y_label = 'Proportion of survived/unsurvived'
               , title = 'Proportion of survived/unsurvived by Pclass (1, 2,3)')
```
