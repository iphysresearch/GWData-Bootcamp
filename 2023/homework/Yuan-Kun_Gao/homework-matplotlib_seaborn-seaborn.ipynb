## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
sns.pairplot(data=data, x_vars=["sepal_length", "sepal_width"], y_vars=["petal_length", "petal_width"])
```

```python
sns.pairplot(data=data, x_vars=["sepal_length", "sepal_width"], y_vars=["petal_length", "petal_width"], hue="species")
```

```python
species = ['setosa', 'versicolor', 'virginica']
data_species = []
for i in species:
    j = data.groupby("species").get_group(i)
    data_species.append(j)
```

```python
index = 0
for i in data_species:
    ax = sns.boxplot(i)
    ax.set_title('The box with 4 features of the %s flowers'%species[index])
    index += 1
    plt.show()
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
data["people"] = [1]*len(data.index)
data
```

```python
current_palette = sns.color_palette()

ax1 = sns.barplot(x="pclass", y="people", data=data)
ax2 = sns.barplot(x="pclass", y="survived", data=data, errorbar=None)
ax2.set_ylabel("proportion")
ax1legend = plt.Rectangle((0,0), 1, 1, fc=current_palette[0], edgecolor="none")
ax2legend = plt.Rectangle((0,0), 1, 1, fc=current_palette[1], edgecolor="none")
plt.legend([ax1legend, ax2legend], ["unsurvived", "survived"], loc="best", bbox_to_anchor=(1, 1))
plt.show()
```

```python
ax1 = sns.barplot(x="sex", y="people", data=data)
ax2 = sns.barplot(x="sex", y="survived", data=data, errorbar=None)
ax2.set_ylabel("proportion")
ax1legend = plt.Rectangle((0,0), 1, 1, fc=current_palette[0], edgecolor="none")
ax2legend = plt.Rectangle((0,0), 1, 1, fc=current_palette[1], edgecolor="none")
plt.legend([ax1legend, ax2legend], ["unsurvived", "survived"], loc="best", bbox_to_anchor=(1, 1))
plt.show()
```
