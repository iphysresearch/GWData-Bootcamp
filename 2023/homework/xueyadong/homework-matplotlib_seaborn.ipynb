## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

sns.scatterplot(x="sepal_length", y="petal_length", data=data, ax=axes[0, 0])
axes[0, 0].set_title("s_l VS p_l")
axes[0, 0].set_xlabel("s_l")
axes[0, 0].set_ylabel("p_l")
sns.scatterplot(x="sepal_length", y="petal_width", data=data, ax=axes[0, 1])
axes[0, 1].set_title("s_l VS p_w")
axes[0, 1].set_xlabel("s_l")
axes[0, 1].set_ylabel("p_w")
sns.scatterplot(x="sepal_width", y="petal_length", data=data, ax=axes[1, 0])
axes[1, 0].set_title("s_w VS p_l")
axes[1, 0].set_xlabel("s_w")
axes[1, 0].set_ylabel("p_l")
sns.scatterplot(x="sepal_width", y="petal_width", data=data, ax=axes[1, 1])
axes[1, 1].set_title("s_w VS p_w")
axes[1, 1].set_xlabel("s_w")
axes[1, 1].set_ylabel("p_w")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

sns.scatterplot(x="sepal_length", y="petal_length", data=data, ax=axes[0, 0], hue="species",legend=None)
axes[0, 0].set_title("s_l VS p_l")
sns.scatterplot(x="sepal_length", y="petal_width", data=data, ax=axes[0, 1], hue="species", legend=None)
axes[0, 1].set_title("s_l VS p_w")
sns.scatterplot(x="sepal_width", y="petal_length", data=data, ax=axes[1, 0], hue="species", legend=None)
axes[1, 0].set_title("s_w VS p_l")
sns.scatterplot(x="sepal_width", y="petal_width", data=data, ax=axes[1, 1], hue="species", legend=None)
axes[1, 1].set_title("s_w VS p_w")

plt.tight_layout()
plt.show()
```

```python
species_list = data["species"].unique()

for species in species_list:
    chosen_data = data[data["species"] == species]

    plt.figure(figsize=(8, 6))
    sns.boxplot(data=chosen_data[["sepal_length", "sepal_width", "petal_length", "petal_width"]])
    plt.title(f"The box with 4 features for the {species} flowers")
    plt.show()
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
data['pclass'] = data['pclass'].astype(str)
data['survived'] = data['survived'].astype(str)
data['sex'] = data['sex'].astype(str)
data["people"] = [1]*len(data.index)
data.head()
```

```python
survival_counts = data.groupby(["pclass", "survived"]).size().unstack()
survival_percentages = survival_counts.div(survival_counts.sum(axis=1), axis=0)
cumulative_percentages = survival_percentages.cumsum(axis=1)

fig, axes = plt.subplots(figsize=(8,6))

for col in cumulative_percentages.columns:
    axes.bar(cumulative_percentages.index, cumulative_percentages[col], label=f'{col} Survived', alpha=0.5)

axes.set_title("Proportion of survived/unsurvived by Pclass (1, 2, 3)")
axes.set_xlabel("Pclass")
axes.set_ylabel("Portion")
plt.show()
```

```python
survival_counts_sex = data.groupby(["sex", "survived"]).size().unstack(fill_value=0)
survival_percentages_sex = survival_counts_sex.div(survival_counts_sex.sum(axis=1), axis=0)
cumulative_percentages_sex = survival_percentages_sex.cumsum(axis=1)
df_sex = cumulative_percentages_sex.stack().reset_index(name='Percentage')

fig, axes = plt.subplots(figsize=(8,6))

for col in cumulative_percentages_sex.columns:
    axes.bar(cumulative_percentages_sex.index, cumulative_percentages_sex[col], label=f'{col} Survived', alpha=0.5)

axes.set_title("Proportion of survived/unsurvived by Sex")
axes.set_xlabel("Sex")
axes.set_ylabel("Portion")
plt.show()
```
