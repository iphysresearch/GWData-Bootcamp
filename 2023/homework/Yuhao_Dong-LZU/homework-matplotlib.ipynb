# 数据可视化作业题目

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rc("font",family=['YouYuan', 'Times New Roman']) #设置中文图例
```

## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
annual_passengers = data.groupby('year')['passengers'].sum() # 按照'year'（年份）对数据进行分组，并对每年的乘客数量进行求和。

plt.figure(figsize=(5, 3), dpi=200) #设置图片大小
annual_passengers.plot(kind='line', marker='o')
plt.title('年度乘客总量变化情况 (1949-1960)')
plt.xlabel('年份')
plt.ylabel('乘客总量')
plt.grid(True) # 显示网格
plt.show()
```

```python
monthly_passengers = data.groupby('month')['passengers'].sum() # 按照'month'（年份）对数据进行分组，并对每月的乘客数量进行求和。

plt.figure(figsize=(6, 3), dpi=200)
# plt.bar(monthly_passengers['month'], monthly_passengers['passengers'], color='skyblue')
monthly_passengers.plot(kind='bar', color='skyblue')
plt.title('The number of passengers in each month')
plt.xlabel('Month')
plt.ylabel('Number')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10), dpi=200) #设置2×2子图

axs[0,0].scatter(data['sepal_length'], data['petal_length'], alpha=0.7)
axs[0,0].set_title('s_l VS p_l')
axs[0,0].set_xlabel('sepal length')
axs[0,0].set_ylabel('petal length')
# axs[0,0].legend()
# axs[0,0].grid(True)

axs[0,1].scatter(data['sepal_length'], data['petal_width'], alpha=0.7)
axs[0,1].set_title('s_l VS p_w')
axs[0,1].set_xlabel('sepal length')
axs[0,1].set_ylabel('petal width')
# axs[0,1].legend()
# axs[0,0].grid(True)

axs[1,0].scatter(data['sepal_width'], data['petal_length'], alpha=0.7)
axs[1,0].set_title('s_w VS p_l')
axs[1,0].set_xlabel('sepal width')
axs[1,0].set_ylabel('petal length')
# axs[1,0].legend()
# axs[0,0].grid(True)

axs[1,1].scatter(data['sepal_width'], data['petal_width'], alpha=0.7)
axs[1,1].set_title('s_w VS p_w')
axs[1,1].set_xlabel('sepal width')
axs[1,1].set_ylabel('petal width')
# axs[1,1].legend()
# axs[0,0].grid(True)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10), dpi=200) #设置2×2子图

species_colors = {
    'setosa': 'red',
    'versicolor': 'green',
    'virginica': 'blue'
} #定义不同种类的散点颜色


for species, color in species_colors.items():
    subset = data[data['species'] == species]
    axs[0,0].scatter(subset['sepal_length'], subset['petal_length'], color = color, alpha=0.7)
axs[0,0].set_title('s_l VS p_l')
axs[0,0].set_xlabel('sepal length')
axs[0,0].set_ylabel('petal length')
# axs[0,0].legend()
# axs[0,0].grid(True)


for species, color in species_colors.items():
    subset = data[data['species'] == species]
    axs[0,1].scatter(subset['sepal_length'], subset['petal_width'],color = color, alpha=0.7)
axs[0,1].set_title('s_l VS p_w')
axs[0,1].set_xlabel('sepal length')
axs[0,1].set_ylabel('petal width')
# axs[0,1].legend()
# axs[0,0].grid(True)


for species, color in species_colors.items():
    subset = data[data['species'] == species]
    axs[1,0].scatter(subset['sepal_width'], subset['petal_length'], color = color,alpha=0.7)
axs[1,0].set_title('s_w VS p_l')
axs[1,0].set_xlabel('sepal width')
axs[1,0].set_ylabel('petal length')
# axs[1,0].legend()
# axs[0,0].grid(True)


for species, color in species_colors.items():
    subset = data[data['species'] == species]
    axs[1,1].scatter(subset['sepal_width'], subset['petal_width'],color = color, alpha=0.7)
axs[1,1].set_title('s_w VS p_w')
axs[1,1].set_xlabel('sepal width')
axs[1,1].set_ylabel('petal width')
# axs[1,1].legend()
# axs[0,0].grid(True)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
subset = data[data['species'] == 'setosa']

# 创建新的数据框，包含需要绘制的数据
plot_data = pd.DataFrame({
    'Sepal Length': subset['sepal_length'],
    'Sepal Width': subset['sepal_width'],
    'Petal Length': subset['petal_length'],
    'Petal Width': subset['petal_width']
})

# 绘制箱线图
ax = plt.figure(figsize=(10, 6), dpi= 200)
plt.boxplot(x=plot_data, data=plot_data)
plt.title('鸢尾花种类 Setosa 的四个特征箱线图')
plt.ylabel('Values')
plt.xticks([1,2,3,4],['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
plt.show()
```

```python
subset = data[data['species'] == 'versicolor']

# 创建新的数据框，包含需要绘制的数据
plot_data = pd.DataFrame({
    'Sepal Length': subset['sepal_length'],
    'Sepal Width': subset['sepal_width'],
    'Petal Length': subset['petal_length'],
    'Petal Width': subset['petal_width']
})

# 绘制箱线图
plt.figure(figsize=(10, 6), dpi= 200)
plt.boxplot(x=plot_data, data=plot_data)
plt.title('鸢尾花种类 Versicolor 的四个特征箱线图')
plt.ylabel('Values')
plt.xticks([1,2,3,4],['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
plt.show()
```

```python
subset = data[data['species'] == 'virginica']

# 创建新的数据框，包含需要绘制的数据
plot_data = pd.DataFrame({
    'Sepal Length': subset['sepal_length'],
    'Sepal Width': subset['sepal_width'],
    'Petal Length': subset['petal_length'],
    'Petal Width': subset['petal_width']
})

# 绘制箱线图
plt.figure(figsize=(10, 6), dpi= 200)
plt.boxplot(x=plot_data, data=plot_data)
plt.title('鸢尾花种类 Virginica 的四个特征箱线图')
plt.ylabel('Values')
plt.xticks([1,2,3,4],['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
plt.show()
```

```python
# subset = data[data['species'] == 'setosa']
species = data['species'].unique()

plt.figure(figsize=(15, 10))

for i, specie in enumerate(species, 1):
    plt.subplot(2, 3, i)
    sns.kdeplot(
        data=data[data['species'] == specie],
        x="sepal_length",
        y="sepal_width",
        fill=True
    )
    plt.title(f"{specie.capitalize()} Sepal Size Distribution")

    plt.subplot(2, 3, i + 3)
    sns.kdeplot(
        data=data[data['species'] == specie],
        x="petal_length",
        y="petal_width",
        fill=True
    )
    plt.title(f"{specie.capitalize()} Petal Size Distribution")

plt.tight_layout()
plt.show()
```

## 练习3：餐厅小费情况分析 (7个题)

1. 小费和总消费之间的关系（提示：散点图+回归分析）
2. 男性顾客和女性顾客，谁更慷慨（提示：箱图或者提琴图）
3. 抽烟与否是否会对小费金额产生影响（提示：箱图或者提琴图）
4. 工作日和周末，什么时候顾客给的小费更慷慨（提示：箱图或者提琴图）
5. 午饭和晚饭，哪一顿顾客更愿意给小费（提示：箱图或者提琴图）
6. 就餐人数是否会对慷慨度产生影响（提示：箱图或者提琴图）
7. 性别+抽烟的组合因素对慷慨度的影响（提示：统计柱状图）

```python
data = sns.load_dataset("tips")
data.head()
# 总消费，小费，性别，吸烟与否，就餐星期，就餐时间，就餐人数
```

```python
# 使用matplotlib绘制散点图
plt.figure(figsize=(8, 6))
plt.scatter(x="total_bill", y="tip", data= data)
plt.title(" Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
male_tips = data[data['sex'] == 'Male']['tip'] #男性小费
female_tips = data[data['sex'] == 'Female']['tip'] #女性小费

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([male_tips, female_tips], labels=['Male', 'Female'])
plt.title("The box with 2 features for the tip")
plt.xlabel("Sex")
plt.ylabel("Tip")
plt.show()
```

```python
yessmoke_tips = data[data['smoker'] == 'Yes']['tip']
nosmoke_tips = data[data['smoker'] == 'No']['tip']

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([yessmoke_tips, nosmoke_tips], labels=['Yes', 'No'])
plt.title("The box with 2 features for the tip")
plt.xlabel("Smoker")
plt.ylabel("Tip")
plt.show()
```

```python
data['Day Type'] = data['day'].apply(lambda x: 'Weekend' if x in ['Sat', 'Sun'] else 'Weekday')

weekend_tips = data[data['Day Type'] == 'Weekend']['tip'] #周末小费
weekday_tips = data[data['Day Type'] == 'Weekday']['tip'] #工作日小费

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([weekday_tips, weekend_tips], labels=['Weekday', 'Weekend'])
plt.title("The box with 2 features for the tip")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
dinner_tips = data[data['time'] == 'Dinner']['tip'] #晚餐小费
lunch_tips = data[data['time'] == 'Lunch']['tip']  #午餐小费

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([dinner_tips, lunch_tips], labels=['Dinner', 'Lunch'])
plt.title("The box with 2 features for the tip")
plt.xlabel("Time")
plt.ylabel("Tip")
plt.show()
```

```python
one_tips = data[data['size'] == 1]['tip']    #一人就餐小费
two_tips = data[data['size'] == 2]['tip']    #两人就餐小费
three_tips = data[data['size'] == 3]['tip']  #三人就餐小费
four_tips = data[data['size'] == 4]['tip']   #四人就餐小费
five_tips = data[data['size'] == 5]['tip']   #五人就餐小费
six_tips = data[data['size'] == 6]['tip']    #六人就餐小费

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([one_tips, two_tips, three_tips, four_tips, five_tips, six_tips], labels=['1', '2', '3', '4', '5', '6'])
plt.title("The box with 6 features for the tip")
plt.xlabel("Size")
plt.ylabel("Tip")
plt.show()
```

```python
data['Sex and Smoker'] = data['sex'].astype(str) + ' - ' + data['smoker'].astype(str)

# 将性别和抽烟数据组合起来
average_tips = data.groupby('Sex and Smoker')['tip'].mean().reset_index()

# 根据小费进行排序
# average_tips = average_tips.sort_values(by='tip', ascending=False)

plt.figure(figsize=(6, 4),dpi = 200)
plt.bar(average_tips['Sex and Smoker'], average_tips['tip'])
plt.title("Values by Sex and Smoker")
plt.xlabel("Sex")
plt.ylabel("Tip")
plt.show()
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
survival_counts = data.groupby(['class', 'survived']).size().unstack(fill_value=0)

# Create the bar plot
classes = survival_counts.index
bar_width = 0.35
index = np.arange(len(classes))

fig, ax = plt.subplots()

bar1 = ax.bar(index, survival_counts[0]/(survival_counts[0] + survival_counts[1]), bar_width, label='Unsurvived', color='cyan')
bar2 = ax.bar(index + bar_width, survival_counts[1]/(survival_counts[0] + survival_counts[1]), bar_width, label='Survived', color='magenta')

ax.set_xlabel('Class')
ax.set_ylabel('Proportion')
ax.set_title('Proportion of Survived and Unsurvived Passengers by Class')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(classes)
ax.legend()

plt.show()
```

```python
survival_counts = data.groupby(['sex', 'survived']).size().unstack(fill_value=0)

# Create the bar plot
classes = survival_counts.index
bar_width = 0.35
index = np.arange(len(classes))

fig, ax = plt.subplots()

bar1 = ax.bar(index, survival_counts[0]/(survival_counts[0] + survival_counts[1]), bar_width, label='Unsurvived', color='cyan')
bar2 = ax.bar(index+ bar_width, survival_counts[1]/(survival_counts[0] + survival_counts[1]), bar_width, label='Survived', color='magenta')

ax.set_xlabel('Sex')
ax.set_ylabel('Proportion')
ax.set_title('Proportion of Survived and Unsurvived Passengers by Sex')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(classes)
ax.legend()

plt.show()
```

```python
unsurvived_fare = data[data['survived'] == 0]['fare'] #未存活的票价
survived_fare = data[data['survived'] == 1]['fare']  #存活的票价

plt.figure(figsize=(6, 4),dpi = 200)
plt.boxplot([unsurvived_fare, survived_fare], labels=['0', '1'])
plt.title("The box with 2 features for the the fare")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```
