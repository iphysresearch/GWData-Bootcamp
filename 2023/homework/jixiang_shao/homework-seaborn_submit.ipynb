## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
data_1  = data[['year','passengers']].groupby('year').agg('sum')
data_1  =  data_1.reset_index()
sns.lineplot(x="year", y="passengers", data=data_1).set_title('The number of passengers in each year')
```

```python
data_2  = data[['month','passengers']].groupby('month').agg('sum')
data_2  = data_2.reset_index()
sns.barplot(data_2['month'],data_2['passengers'],color='lightblue').set_title('The number of passengers in each month')
sns.set_style("darkgrid")
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
x_vars = ['sepal_length','sepal_width']
y_vars = ['petal_length','petal_width']
g=sns.PairGrid(data,x_vars=x_vars, y_vars=y_vars,height=3.0)
g=g.map(sns.scatterplot)
```

```python
x_vars = ['sepal_length','sepal_width']
y_vars = ['petal_length','petal_width']
g=sns.PairGrid(data,hue = 'species', x_vars=x_vars, y_vars=y_vars,height=3.0)
g=g.map(sns.scatterplot)
```

```python
species = list(set(data['species']))
print (species)

for i in range(3):
    plt.figure()
    data_1 = data[data['species'] == species[i]][['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
    sns.boxplot( data = data_1).set_title('The box with 4 features for the ' +str(species[i])+' flowers')
    plt.xlabel('Features of the '+str(species[i])+ ' flowers')
    plt.ylabel('Values')
```

7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
data['unsurvived'] = 1 - data['survived']
data.info()
```

```python
def plot_fig(keyword):
    data_1 =  data.groupby(keyword)[['survived','unsurvived']].agg('sum')

    data_1['total'] = data_1['survived'] + data_1['unsurvived']
    data_1['total_prop'] =1
    data_1['survived_prop'] = data_1['survived']/(data_1['survived'] + data_1['unsurvived'])
    data_1['unsurvived_prop'] = data_1['unsurvived']/(data_1['survived'] + data_1['unsurvived'])
    print (data_1)


    fig = plt.figure(figsize=(6,4))
    plt.rcParams['figure.constrained_layout.use'] = True
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif')
    plt.rcParams['font.size'] = 15 

    ax = fig.add_subplot(111) 
    sns.barplot(data_1, x= str(keyword), y='total_prop')
    sns.barplot(data_1, x= str(keyword), y='survived_prop',label='survived')
    

plot_fig(keyword='pclass')
```

```python
plot_fig(keyword='sex')
```
