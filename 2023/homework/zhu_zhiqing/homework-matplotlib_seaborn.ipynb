```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# exercise 1
## load data

```python
data = sns.load_dataset("flights")
data.head()
```

## Problem 1

```python
year_passengers = data['passengers'].groupby(data['year']).sum()
sns.set()
plt.plot(year_passengers.index,year_passengers)
plt.xticks(year_passengers.index)
plt.xlabel('year')
plt.ylabel('number')
plt.title('The number of passengers in each year');
```

## Problem 2

```python
month_passengers = data['passengers'].groupby(data['month'],observed=False).sum()
plt.bar(month_passengers.index,month_passengers)
plt.xticks(month_passengers.index)
plt.xlabel('month')
plt.ylabel('number')
plt.title('The number of passengers in each month');
```

# exercise 2
## load data

```python
data2 = sns.load_dataset("iris")
data2
```

## Problem 1

```python
# plot 1
plt.subplot(2,2,1)
plt.scatter(data2['sepal_length'],data2['petal_length'],s=10)
plt.xlabel('s_l',fontsize=7)
plt.ylabel('p_l',fontsize=7)
plt.title('s_l VS p_l',fontsize=10)

# plot 2
plt.subplot(2,2,2)
plt.scatter(data2['sepal_length'],data2['petal_width'],s=10)
plt.xlabel('s_l',fontsize=7)
plt.ylabel('p_w',fontsize=7)
plt.title('s_l VS p_w',fontsize=10)

# plot 3
plt.subplot(2,2,3)
plt.scatter(data2['sepal_width'],data2['petal_length'],s=10)
plt.xlabel('s_w',fontsize=7)
plt.ylabel('p_l',fontsize=7)
plt.title('s_w VS p_l',fontsize=10)

# plot 4
plt.subplot(2,2,4)
plt.scatter(data2['sepal_width'],data2['petal_width'],s=10)
plt.xlabel('s_w',fontsize=7)
plt.ylabel('p_w',fontsize=7)
plt.title('s_w VS p_w',fontsize=10)

plt.subplots_adjust(wspace=0.2, hspace=0.5)
```

# exercise 4
## load data

```python
data4 = sns.load_dataset("titanic")
data4.head()
```

## Problem 1

```python
sur = data4['survived'].groupby(data4['pclass']).sum()
die = data4['pclass'].value_counts() - sur
bottom = np.zeros(3)
plt.bar(sur.index,(sur/die),bottom=bottom,label='survived')
bottom += sur/die
plt.bar(sur.index,die/die,bottom=bottom,label='died')
plt.legend()
```

## Problem 2

```python
sur1 = data4['survived'].groupby(data4['sex']).sum()
die1 = data4['sex'].value_counts() - sur1
bottom = np.zeros(2)
plt.bar(sur1.index,(sur1/die1),bottom=bottom,label='survived')
bottom += sur1/die1
plt.bar(sur1.index,die1/die1,bottom=bottom,label='died')
plt.legend()
```

## Problem 3

```python
plt.boxplot([data4[data4['survived']==1]['fare'],data4[data4['survived']==0]['fare']],labels=['survived','died']);
plt.ylabel('fare')
```

## Problem 4

```python
plt.boxplot([data4[data4['survived']==1]['age'].dropna(),data4[data4['survived']==0]['age'].dropna()],labels=['survived','died']);
plt.ylabel('age')
```

## Problem 5

```python
table = data4.groupby(['embarked','pclass']).size().reset_index(name='Count')
print(table)
```

```python
x = np.arange(3)
width = 0.25
plt.bar(x-width,table[table['embarked']=='C']['Count'],width,label='C')
plt.bar(x,table[table['embarked']=='Q']['Count'],width,label='Q')
plt.bar(x+width,table[table['embarked']=='S']['Count'],width,label='S')
plt.legend()
plt.xticks(x, [1,2,3])
```

## Problem 6

```python
plt.boxplot([data4[data4['survived']==1]['sibsp'].dropna(),data4[data4['survived']==0]['sibsp'].dropna()],labels=['survived','died']);
plt.ylabel('sibsp')
```

## Problem 7

```python
plt.boxplot([data4[data4['survived']==1]['parch'].dropna(),data4[data4['survived']==0]['parch'].dropna()],labels=['survived','died']);
plt.ylabel('parch')
```
