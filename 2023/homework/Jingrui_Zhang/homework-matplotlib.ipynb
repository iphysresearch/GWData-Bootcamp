## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
passengers_in_each_year = data.drop('month',axis=1).groupby('year').sum()['passengers']
years_menu = passengers_in_each_year.keys()
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("The number of passengers in each year")
plt.xlabel("year")
plt.ylabel("number")
plt.plot(years_menu,passengers_in_each_year)

passengers_in_each_month = data.drop('year',axis=1).groupby('month').sum()['passengers']
months_menu = passengers_in_each_month.keys()
plt.subplot(1,2,2)
plt.title("The number of passengers in each month")
plt.xlabel("month")
plt.ylabel("number")
plt.bar(months_menu,passengers_in_each_month)
plt.show()
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
survived_no_pclass = data.groupby("pclass")['survived'].agg('sum')
unservived_no_pclass = data.groupby("pclass")['survived'].agg('count')-survived_no_pclass
total_no_pclass = data.groupby("pclass")['survived'].agg('count')
survived_prop_pclass = survived_no_pclass/total_no_pclass
unsurvived_prop_pclass = unservived_no_pclass/total_no_pclass

df_pclass = pd.DataFrame({
    'survived':survived_no_pclass,
    'unsurvived':unservived_no_pclass,
    'total_no':total_no_pclass,
    'survived_prop':survived_prop_pclass,
    'unsurvived_prop':unsurvived_prop_pclass,
})
print(df_pclass)

plt.figure()
plt.title('Proportion of survived/unsurvived by Pclass (1,2,3)')
plt.xlabel('Pclass')
plt.ylabel('Proportion of survived/unsurvived')
plt.bar(df_pclass.index,df_pclass['survived_prop'],label = 'Survived',width = 0.7)
plt.bar(df_pclass.index,df_pclass['unsurvived_prop'],bottom = df_pclass['survived_prop'],label = 'Unsurvived',width = 0.7)
plt.legend(bbox_to_anchor=(1.25, 1.15))
plt.xticks([1,2,3],['1','2','3'])
plt.show()
```

```python
survived_no_sex = data.groupby("sex")['survived'].agg('sum')
unservived_no_sex = data.groupby("sex")['survived'].agg('count')-survived_no_sex
total_no_sex = data.groupby("sex")['survived'].agg('count')
survived_prop_sex = survived_no_sex/total_no_sex
unsurvived_prop_sex = unservived_no_sex/total_no_sex

df_sex = pd.DataFrame({
    'survived':survived_no_sex,
    'unsurvived':unservived_no_sex,
    'total_no':total_no_sex,
    'survived_prop':survived_prop_sex,
    'unsurvived_prop':unsurvived_prop_sex,
})
print(df_sex)

plt.figure()
plt.title('Proportion of survived/unsurvived by sex')
plt.xlabel('Sex')
plt.ylabel('Proportion of survived/unsurvived')
plt.bar(df_sex.index,df_sex['survived_prop'],label = 'Survived',width = 0.7)
plt.bar(df_sex.index,df_sex['unsurvived_prop'],bottom = df_sex['survived_prop'],label = 'Unsurvived',width = 0.7)
plt.legend(bbox_to_anchor=(1.25, 1.15))
plt.show()
```
