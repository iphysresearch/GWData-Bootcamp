```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python

# 年度乘客总量的变化情况（折线图）
plt.figure(figsize=(10, 6))
total_passengers_by_year = data.groupby('year')['passengers'].sum()
plt.plot(total_passengers_by_year.index, total_passengers_by_year.values, marker='o')
plt.title('Yearly Total Passengers')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.grid(True)
plt.show()
```

```python
plt.bar(data['month'], data['passengers'])
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Passenger Distribution Across Months')
plt.show()
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
display(data['species'].unique())
```

```python
sepal_length = data['sepal_length']
petal_length = data['petal_length']
sepal_width = data['sepal_width']
petal_width	= data['petal_width']
species = data['species']
# 设置图的布局
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
# 1. 萼片和花瓣大小关系的散点图
axs[0, 0].scatter(sepal_length, petal_length, alpha=0.8)
axs[0, 0].set_title('萼片和花瓣长度关系的散点图')
axs[0, 0].set_xlabel('萼片长度')
axs[0, 0].set_ylabel('花瓣长度')

axs[0, 1].scatter(sepal_width, petal_width,alpha=0.8)
axs[0, 1].set_title('萼片和花瓣宽度关系的散点图')
axs[0, 1].set_xlabel('萼片宽度')
axs[0, 1].set_ylabel('花瓣宽度')

axs[1, 0].scatter(sepal_width, petal_length,alpha=0.8)
axs[1, 0].set_title('萼片宽度和花瓣长度关系的散点图')
axs[1, 0].set_xlabel('萼片宽度')
axs[1, 0].set_ylabel('花瓣长度')


axs[1, 1].scatter(sepal_length, petal_width,alpha=0.8)
axs[1, 1].set_title('萼片长度和花瓣宽度关系的散点图')
axs[1, 1].set_xlabel('萼片长度')
axs[1, 1].set_ylabel('花瓣宽度')

# 调整子图之间的间距
plt.tight_layout()
plt.show()
```

```python
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

# 设置图的布局
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. 萼片和花瓣大小关系的散点图
axs[0, 0].scatter(sepal_length, petal_length, c=data['species'].map(colors), alpha=0.8)
axs[0, 0].set_title('萼片和花瓣长度关系的散点图')
axs[0, 0].set_xlabel('萼片长度')
axs[0, 0].set_ylabel('花瓣长度')

# 2. 萼片和花瓣宽度关系的散点图
axs[0, 1].scatter(sepal_width, petal_width, c=data['species'].map(colors), alpha=0.8)
axs[0, 1].set_title('萼片和花瓣宽度关系的散点图')
axs[0, 1].set_xlabel('萼片宽度')
axs[0, 1].set_ylabel('花瓣宽度')

# 3. 萼片宽度和花瓣长度关系的散点图
axs[1, 0].scatter(sepal_width, petal_length, c=data['species'].map(colors), alpha=0.8)
axs[1, 0].set_title('萼片宽度和花瓣长度关系的散点图')
axs[1, 0].set_xlabel('萼片宽度')
axs[1, 0].set_ylabel('花瓣长度')

# 4. 萼片长度和花瓣宽度关系的散点图
axs[1, 1].scatter(sepal_length, petal_width, c=data['species'].map(colors), alpha=0.8)
axs[1, 1].set_title('萼片长度和花瓣宽度关系的散点图')
axs[1, 1].set_xlabel('萼片长度')
axs[1, 1].set_ylabel('花瓣宽度')

# 调整子图之间的间距
plt.tight_layout()
plt.show()
```
