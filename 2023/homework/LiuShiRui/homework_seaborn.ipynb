```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("flights")
data.head()
```

```python
f = data.groupby('year').sum(numeric_only=True)
f = f.reset_index()

sns.lineplot(data=f,x='year',y='passengers')
plt.title("The number of passengers in each year")
```

```python
sns.barplot(data,x = "month", y = "passengers",estimator='sum',errorbar=None)
plt.title("The number of passsengers in each month")
```
