# 1. 航班乘客变化

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
data = sns.load_dataset("flights")
yearSum=data.groupby('year')['passengers'].agg('sum').reset_index()
monthSum=data.groupby('month')['passengers'].agg('sum').reset_index()
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(yearSum['year'],yearSum['passengers'])
plt.xlabel('year')
plt.ylabel('numbers')
plt.title('the number of passengers ineach year')
plt.subplot(1,2,2)
plt.bar(np.arange(1,13),monthSum['passengers'])
plt.xlabel('month')
plt.xticks(np.arange(1,13,1))
plt.ylabel('numbers')
plt.title('the number of passengers ineach month')
plt.show()
```

# 2. 鸢尾花花型尺寸分析
## 萼片（sepal）和花瓣（petal）的大小关系

```python
data = sns.load_dataset("iris")
data.set_index(["species"],inplace=True)
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.scatter(data['sepal_length'],data['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.subplot(2,2,2)
plt.scatter(data['sepal_length'],data['petal_width'])
plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.subplot(2,2,3)
plt.scatter(data['sepal_width'],data['petal_length'])
plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.subplot(2,2,4)
plt.scatter(data['sepal_width'],data['petal_width'])
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.show()
```

## 不同种类（species）鸢尾花sepal和petal的大小关系

```python
plt.figure()
plt.boxplot([data.loc['setosa','sepal_length'],data.loc['setosa','sepal_width'],\
             data.loc['setosa','petal_length'],data.loc['setosa','petal_width']],\
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.xlabel('features of setosa flower')
plt.ylabel('values')
plt.title('the box with 4 features for setosa flower')
plt.show()
plt.figure()
plt.boxplot([data.loc['versicolor','sepal_length'],data.loc['versicolor','sepal_width'],\
             data.loc['versicolor','petal_length'],data.loc['versicolor','petal_width']],\
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.xlabel('features of versicolor flower')
plt.ylabel('values')
plt.title('the box with 4 features for versicolor flower')
plt.show()
plt.figure()
plt.boxplot([data.loc['virginica','sepal_length'],data.loc['virginica','sepal_width'],\
             data.loc['virginica','petal_length'],data.loc['virginica','petal_width']],\
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.xlabel('features of virginica flower')
plt.ylabel('values')
plt.title('the box with 4 features for virginica flower')
plt.show()
```

## 不同种类鸢尾花萼片和花瓣大小的分布情况

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.scatter(data.loc['setosa','sepal_length'],data.loc['setosa','petal_length'])
plt.scatter(data.loc['versicolor','sepal_length'],data.loc['versicolor','petal_length'])
plt.scatter(data.loc['virginica','sepal_length'],data.loc['virginica','petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.subplot(2,2,2)
plt.scatter(data.loc['setosa','sepal_length'],data.loc['setosa','petal_width'])
plt.scatter(data.loc['versicolor','sepal_length'],data.loc['versicolor','petal_width'])
plt.scatter(data.loc['virginica','sepal_length'],data.loc['virginica','petal_width'])
plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.subplot(2,2,3)
plt.scatter(data.loc['setosa','sepal_width'],data.loc['setosa','petal_length'])
plt.scatter(data.loc['versicolor','sepal_width'],data.loc['versicolor','petal_length'])
plt.scatter(data.loc['virginica','sepal_width'],data.loc['virginica','petal_length'])
plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.subplot(2,2,4)
plt.scatter(data.loc['setosa','sepal_width'],data.loc['setosa','petal_width'])
plt.scatter(data.loc['versicolor','sepal_width'],data.loc['versicolor','petal_width'])
plt.scatter(data.loc['virginica','sepal_width'],data.loc['virginica','petal_width'])
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.show()
```

# 3 餐厅小费情况分析
## 小费和总消费之间的关系

```python
data = sns.load_dataset("tips")
plt.figure()
plt.scatter(data['total_bill'],data['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('total_bill vs. tip')
plt.show()
```

## 男性顾客和女性顾客，谁更慷慨

```python
data.set_index(['sex'],inplace=True)
plt.figure()
plt.boxplot([data.loc['Male','tip'],data.loc['Female','tip']],labels=['Male','Female'])
plt.xlabel('features of sex')
plt.ylabel('values of tip')
plt.title('the box with 2 features for the tip')
plt.show()
data=data.reset_index()
```

## 抽烟与否是否会对小费金额产生影响

```python
data.set_index(['smoker'],inplace=True)
plt.figure()
plt.boxplot([data.loc['Yes','tip'],data.loc['No','tip']],labels=['Yes','No'])
plt.xlabel('features of smoker')
plt.ylabel('values of tip')
plt.title('the box with 2 features for the tip')
plt.show()
data=data.reset_index()
```

## 工作日和周末，什么时候顾客给的小费更慷慨

```python
data.set_index(['day'],inplace=True)
plt.figure()
plt.boxplot([data.loc['Sun','tip'],data.loc['Sat','tip'],data.loc['Thur','tip'],data.loc['Fri','tip']],labels=['Sun','Sat','Thur','Fri'])
plt.xlabel('features of the day')
plt.ylabel('values of tip')
plt.title('the box with 4 features for the tip')
plt.show()
data=data.reset_index()
```

## 午饭和晚饭，哪一顿顾客更愿意给小费

```python
data.set_index(['time'],inplace=True)
plt.figure()
plt.boxplot([data.loc['Dinner','tip'],data.loc['Lunch','tip']],labels=['Dinner','Lunch'])
plt.xlabel('features of time')
plt.ylabel('values of tip')
plt.title('the box with 2 features for the tip')
plt.show()
data=data.reset_index()
```

## 就餐人数是否会对慷慨度产生影响

```python
data.set_index(['size'],inplace=True)
plt.figure()
plt.boxplot([data.loc[1,'tip'],data.loc[2,'tip'],data.loc[3,'tip'],\
           data.loc[4,'tip'],data.loc[5,'tip'],data.loc[6,'tip']],labels=['1','2','3','4','5','6'])
plt.xlabel('features of size')
plt.ylabel('values of tip')
plt.title('')
plt.show('the bos with 6 features for the tip')
data=data.reset_index()
```

## 性别+抽烟的组合因素对慷慨度的影响

```python
sex_smoker=data.groupby(['sex','smoker'])['tip'].agg('mean').unstack()
plt.figure()
plt.bar([0.9,1.4],sex_smoker['Yes'],width=0.2,label='yes',)
plt.bar([1.1,1.6],sex_smoker['No'],width=0.2,label='no')
plt.xticks([1,1.5],['Male','Female'])
plt.legend()
plt.xlabel('sex')
plt.ylabel('mean value of tip')
plt.title('values by sex (Male or Female) and smoker (Yes or No)')
plt.show()
```

# 4 泰坦尼克号海难幸存状况分析 
## 不同仓位等级中幸存和遇难的乘客比例

```python
data = sns.load_dataset("titanic")
TotNum=data['pclass'].value_counts()
SurviveNum=data.groupby(['pclass'])['survived'].agg('sum')
data1=pd.concat([TotNum,SurviveNum],axis=1).reset_index()
data1.columns=['pclass','total','survived']
data1['unsurvived']=data1['total']-data1['survived']
data1['survived_prop']=data1['survived']/data1['unsurvived']
data1['unsurvived_prop']=data1['unsurvived']/data1['unsurvived']
plt.figure()
plt.bar(data1['pclass'],data1['survived_prop'],label='survived')
plt.bar(data1['pclass'],data1['unsurvived_prop'],bottom=data1['survived_prop'],label='unsurvived')
plt.xticks([1,2,3],['1','2','3'])
plt.xlabel('pclass')
plt.ylabel('pproportion of survived/unsurvived')
plt.legend()
plt.title('pproportion of survived/unsurvived by pclass(1,2,3)')
plt.show()
```

## 不同性别的幸存比例

```python
data2=pd.concat([data.groupby('sex')['survived'].agg('sum'),data['sex'].value_counts()],axis=1).reset_index()
data2.columns=['sex','survived','total']
data2['unsurvived']=data2['total']-data2['survived']
data2['survived_prop']=data2['survived']/data2['unsurvived']
data2['unsurvived_prop']=data2['unsurvived']/data2['unsurvived']
plt.figure()
plt.bar(data2['sex'],data2['survived_prop'],label='survived')
plt.bar(data2['sex'],data2['unsurvived_prop'],bottom=data2['survived_prop'],label='unsurvived')
#plt.xticks([1,2],['female','male'])
plt.xlabel('sex')
plt.ylabel('pproportion of survived/unsurvived')
plt.legend()
plt.title('pproportion of survived/unsurvived by sex(male or female)')
plt.show()
```

## 幸存和遇难乘客的票价分布

```python
plt.figure()
plt.boxplot([data.loc[data['survived']==0]['fare'],data.loc[data['survived']==1]['fare']],labels=['0','1'])
plt.xlabel('features of survived')
plt.ylabel('values of fare')
plt.title('the box with 2 features for the fare')
plt.show()
```

## 幸存和遇难乘客的年龄分布

```python
data4=data.dropna(how='any',axis=0,subset='age')
plt.figure()
plt.boxplot([data4.loc[data4['survived']==0]['age'],data4.loc[data4['survived']==1]['age']],labels=['0','1'])
plt.xlabel('features of survived')
plt.ylabel('values of age')
plt.title('the box with 2 features for the age')
plt.show()
```

## 不同上船港口的乘客仓位等级分布

```python
data5=data.groupby(['pclass'])['embarked'].value_counts().unstack()
x=np.arange(1,4,1)
width=0.2
plt.figure()
plt.bar(x,data5['C'],label='C',width=width)
plt.bar(x+width,data5['Q'],label='Q',width=width)
plt.bar(x+2*width,data5['S'],label='S',width=width)
plt.xticks(x+width,['1','2','3'])
plt.legend()
plt.xlabel('pclass')
plt.ylabel('Numbers of people for different kinds of emarked')
plt.title('Numbers of people for different kinds of emarked by pclass')
plt.show()
```

## 幸存和遇难乘客堂兄弟姐妹的数量分布

```python
plt.figure()
plt.boxplot([data.loc[data['survived']==0,'sibsp'],\
            data.loc[data['survived']==1,'sibsp']],\
           labels=['0','1'])
plt.xlabel('features of survived')
plt.ylabel('values of sibsp')
plt.title('the box with 2 features for the sibsp')
plt.show()
```

## 幸存和遇难乘客父母子女的数量分布

```python
plt.figure()
plt.boxplot([data.loc[data['survived']==0,'parch'],\
            data.loc[data['survived']==1,'parch']],\
           labels=['0','1'])
plt.xlabel('features of survived')
plt.ylabel('values of parch')
plt.title('the box with 2 features for the parch')
plt.show()
```

## 单独乘船与否和幸存之间有没有联系

```python
data8=data.groupby(['alone'])['survived'].value_counts().unstack()
x=np.arange(1,3,1)
width=0.4
plt.figure()
plt.bar(x,data8[0],width=width,label='0')
plt.bar(x+width,data8[1],width=width,label='1')
plt.xticks(x+width/2,['False','True'])
plt.legend()
plt.show()
```
