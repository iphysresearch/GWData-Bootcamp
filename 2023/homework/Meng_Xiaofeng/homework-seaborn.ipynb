#### 练习1: 航班乘客变化分析
1. 分析年度乘客总量的变化情况
2. 分析乘客量在一年中12个月份的分布

```python
data = sns.load_dataset("flights") # 导航班乘客数据
data.head()
# 年份, 月份, 乘客数
```

* 练习1  问题1  折线图

```python
'''练习1  问题1  折线图'''
# 风格管理
sns.set_style("ticks")
sns.set_context("notebook")

# 绘图
sns.lineplot(
    data=data, x="year", y="passengers", estimator="sum", errorbar=None,
).set_title("The number of passengers in each year")

plt.xlabel("year") # 设置横坐标标题
plt.ylabel("number") # 设置纵坐标标题

plt.show()
```

* 练习1  问题2  柱状图

```python
'''练习1  问题2  柱状图'''
# 风格管理
sns.set_style("ticks")
sns.set_context("notebook")

# 绘图
sns.barplot(
    data=data, x="month", y="passengers", 
    estimator="sum", errorbar=None,
    ).set_title("The number of passengers in each month")

plt.xlabel("month") # 设置横坐标标题
plt.ylabel("number") # 设置纵坐标标题

plt.show()
```

#### 练习2: 鸢尾花花型尺寸分析
1. 萼片 (sepal) 和花瓣 (petal) 的大小关系  
2. 不同种类 (species) 鸢尾花萼片和花瓣的大小关系  
3. 不同种类鸢尾花萼片和花瓣大小的分布情况

```python
data = sns.load_dataset("iris") # 导入鸢尾花数据
data.head() 
# 萼片长度, 萼片宽度, 花瓣长度, 花瓣宽度, 种类
```

* 练习2  问题1  散点图

```python
'''练习2  问题1  散点图'''
## 定义绘图函数
def PlotScatter(X, Y, Title, indexs):
    # 风格管理
    sns.set_style("ticks")
    sns.set_context("notebook")
    
    # 绘图
    plt.subplot(2, 2, indexs)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    sns.scatterplot(
        data=data, x=X, y=Y,
        ).set_title(Title)


fig, ax = plt.subplots()
fig.set_size_inches(9, 9) # 设置图像大小
plt.axis('off') # 去掉坐标轴


PlotScatter("sepal_length", "petal_length", "s_l vs. p_l", 1) # s_l vs. p_l
PlotScatter("sepal_length", "petal_width", "s_l vs. p_l", 2) # s_l vs. p_w
PlotScatter("sepal_width", "petal_length", "s_w vs. p_l", 3) # s_w vs. p_l
PlotScatter("sepal_width", "petal_width", "s_w vs. p_w", 4) # s_w vs. p_w

plt.show()
```

* 练习2  问题2 散点图 区分不同鸢尾花

```python
'''练习2  问题2 散点图 区分不同鸢尾花'''
## 定义绘图函数
def PlotScatter(X, Y, Title, indexs):
    # 风格管理
    sns.set_style("ticks")
    sns.set_context("notebook")
    
    # 绘图
    plt.subplot(2, 2, indexs)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    sns.scatterplot(
        data=data, x=X, y=Y, hue="species"
        ).set_title(Title)
    
fig, ax = plt.subplots()
fig.set_size_inches(9, 9) # 设置图像大小
plt.axis('off') # 去掉坐标轴


PlotScatter("sepal_length", "petal_length", "s_l vs. p_l", 1) # s_l vs. p_l
PlotScatter("sepal_length", "petal_width", "s_l vs. p_l", 2) # s_l vs. p_w
PlotScatter("sepal_width", "petal_length", "s_w vs. p_l", 3) # s_w vs. p_l
PlotScatter("sepal_width", "petal_width", "s_w vs. p_w", 4) # s_w vs. p_w

plt.show()
```

* 练习2  问题3  箱图

```python
'''练习2  问题3  箱图'''
## 定义绘图函数
def PlotBox(Y, indexs):
    ## 风格管理
    sns.set_style("ticks")
    sns.set_context("notebook")
    ## 绘图
    plt.subplot(2, 2, indexs)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    sns.boxplot(data=data, x="species", y=Y, width=0.3
                ).set_title("species and size")
    plt.xlabel("species") # 设置横坐标标题
    plt.ylabel("size") # 设置纵坐标标题
    
    
fig, ax = plt.subplots()
fig.set_size_inches(9, 9) # 设置图像大小
plt.axis('off') # 去掉坐标轴
PlotBox("sepal_length", 1)
PlotBox("sepal_width", 2)
PlotBox("petal_length", 3)
PlotBox("petal_width", 4)

plt.show()
```

#### 练习3: 餐厅小费情况分析
1. 小费与总消费之间的关系
2. 男性顾客和女性顾客, 谁更慷慨
3. 抽烟与否是否会对小费金额产生影响
4. 工作日和周末, 什么时候顾客给的小费更慷慨
5. 午饭和晚饭, 哪一顿顾客更愿意给小费
6. 就餐人数是否会对慷慨度产生影响
7. 性别 + 抽烟的组合因素对慷慨度的影响

```python
data = sns.load_dataset("tips") # 导入餐厅小费数据
data.head()
# 总消费, 小费, 性别, 吸烟与否, 就餐星期, 就餐时间, 就餐人数
```

* 练习3  问题1  散点图 + 回归分析

```python
'''练习3  问题1  散点图 + 回归分析'''
## 风格管理
sns.set_style("ticks")
sns.set_context("notebook")

## 绘图
sns.regplot(
    x="total_bill", y="tip", data=data,
    ).set_title("total_bill vs. tip")

plt.xlabel("total bill") # 设置横坐标标题
plt.ylabel("tip") # 设置纵坐标标题

plt.show()
```

#### 练习4: 泰坦尼克号海难幸存状况分析

4. 幸存和遇难乘客的年龄分布
5. 不同上船港口的乘客舱位等级分布

7. 幸存和遇难乘客父母子女的数量分布

* 练习4  问题1  提琴图
