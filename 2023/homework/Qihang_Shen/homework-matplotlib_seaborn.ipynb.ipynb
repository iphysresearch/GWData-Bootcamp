```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
```

第一题

```python
data=sns.load_dataset("flights")
data.head()
```

matplotlib

```python
year_number=dict()
for i in range(12):
    year_number[str(1949+i)]=data[data.year==1949+i].passengers.sum()
plt.plot(year_number.keys(),year_number.values())
plt.title("The number of passengers in each year")
plt.xlabel("year")
plt.ylabel("number")
```

```python
month_number=data.groupby("month").agg({"passengers":"sum"})
plt.bar(month_number.reset_index().month,month_number.reset_index().passengers)
plt.title("The number of passengers in each month")
plt.xlabel("month")
plt.ylabel("number")
```

seaborn

```python
sns.set_palette(sns.color_palette("husl",10))
sns.lineplot(x="year",y="passengers",data=pd.DataFrame(list(year_number.items()),columns=['year','passengers']))
plt.title("The number of passengers in each year")
```

```python
sns.barplot(x="month",y="passengers",data=month_number)
plt.title("The number of passengers in each month")
```

第二题

```python
data=sns.load_dataset("iris")
data.head()
```

matplotlib

```python
sns.set()
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].scatter(data["sepal_length"],data["petal_length"])
axs[0,0].set_title("s_l VS. p_l")
axs[0,1].scatter(data["sepal_length"],data["petal_width"])
axs[0,1].set_title("s_l VS. p_w")
axs[1,0].scatter(data["sepal_width"],data["petal_length"])
axs[1,0].set_title("s_w VS. p_l")
axs[1,1].scatter(data["sepal_width"],data["petal_width"])
axs[1,1].set_title("s_w VS. p_w")
```

```python
sns.set()
color_map={"setosa":"blue","versicolor":"green","virginica":"red"}
c=data["species"].map(color_map)
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].scatter(data["sepal_length"],data["petal_length"],c=c)
axs[0,0].set_title("s_l VS. p_l")
axs[0,1].scatter(data["sepal_length"],data["petal_width"],c=c)
axs[0,1].set_title("s_l VS. p_w")
axs[1,0].scatter(data["sepal_width"],data["petal_length"],c=c)
axs[1,0].set_title("s_w VS. p_l")
axs[1,1].scatter(data["sepal_width"],data["petal_width"],c=c)
axs[1,1].set_title("s_w VS. p_w")
```

```python
plt.violinplot(data.set_index("species").loc["setosa"])
plt.xticks([1,2,3,4],["sepal_length","sepal_width","petal_length","petal_width"])
plt.title("The violin with 4 features for the setosa flowers")
plt.xlabel("Features of setosa flowers")
plt.ylabel("Values")
plt.figure()
plt.boxplot(data.set_index("species").loc["versicolor"])
plt.xticks([1,2,3,4],["sepal_length","sepal_width","petal_length","petal_width"])
plt.title("The box with 4 features for the versicolor flowers")
plt.xlabel("Features of versicolor flowers")
plt.ylabel("Values")
plt.figure()
plt.boxplot(data.set_index("species").loc["virginica"],labels=["sepal_length","sepal_width","petal_length","petal_width"])
plt.title("The box with 4 features for the virginica flowers")
plt.xlabel("Features of virginica flowers")
plt.ylabel("Values")
```

seaborn

```python
sns.set_palette(sns.color_palette("husl",10))
fig,axs=plt.subplots(2,2,figsize=(10,12))
sns.scatterplot(data[["sepal_length","petal_length"]].set_index("sepal_length"),ax=axs[0,0],legend=False)
axs[0,0].set_title("s_l VS. p_l")
axs[0,0].set_xlabel("s_l")
axs[0,0].set_ylabel("p_l")
sns.scatterplot(data[["sepal_length","petal_width"]].set_index("sepal_length"),ax=axs[0,1],legend=False)
axs[0,1].set_title("s_l VS. p_w")
axs[0,1].set_xlabel("s_l")
axs[0,1].set_ylabel("p_w")
sns.scatterplot(data[["sepal_width","petal_length"]].set_index("sepal_width"),ax=axs[1,0],legend=False)
axs[1,0].set_title("s_w VS. p_l")
axs[1,0].set_xlabel("s_w")
axs[1,0].set_ylabel("p_l")
sns.scatterplot(data[["sepal_width","petal_width"]].set_index("sepal_width"),ax=axs[1,1],legend=False)
axs[1,1].set_title("s_w VS. p_w")
axs[1,1].set_xlabel("s_w")
axs[1,1].set_ylabel("p_w")
```

```python
sns.set_palette(sns.color_palette("husl",10))
fig,axs=plt.subplots(2,2,figsize=(10,12))
sns.scatterplot(x="sepal_length",y="petal_length",data=data,hue="species",ax=axs[0,0],legend=False)
axs[0,0].set_title("s_l VS. p_l")
axs[0,0].set_xlabel("s_l")
axs[0,0].set_ylabel("p_l")
sns.scatterplot(x="sepal_length",y="petal_width",data=data,hue="species",ax=axs[0,1],legend=False)
axs[0,1].set_title("s_l VS. p_w")
axs[0,1].set_xlabel("s_l")
axs[0,1].set_ylabel("p_w")
sns.scatterplot(x="sepal_width",y="petal_length",data=data,hue="species",ax=axs[1,0],legend=False)
axs[1,0].set_title("s_w VS. p_l")
axs[1,0].set_xlabel("s_w")
axs[1,0].set_ylabel("p_l")
sns.scatterplot(x="sepal_width",y="petal_width",data=data,hue="species",ax=axs[1,1],legend=False)
axs[1,1].set_title("s_w VS. p_w")
axs[1,1].set_xlabel("s_w")
axs[1,1].set_ylabel("p_w")
```

```python
sns.violinplot(data.set_index("species").loc["setosa"])
plt.title("The violin with 4 features for the setosa flowers")
plt.xlabel("Features of setosa flowers")
plt.ylabel("Values")
plt.figure()
sns.boxplot(data.set_index("species").loc["versicolor"])
plt.title("The box with 4 features for the versicolor flowers")
plt.xlabel("Features of versicolor flowers")
plt.ylabel("Values")
plt.figure()
sns.boxplot(data.set_index("species").loc["virginica"])
plt.title("The box with 4 features for the virginica flowers")
plt.xlabel("Features of virginica flowers")
plt.ylabel("Values")
```

第三题

```python
data=sns.load_dataset("tips")
data.head()
```

matplotlib

```python
sns.set()
x=data["total_bill"].values
y=data["tip"].values
plt.scatter(x,y)
plt.title("total bill VS. tip")
plt.xlabel("total bill")
plt.ylabel("tip")
(k,t,r,p,s)=stats.linregress(x,y)
l=k*x+t
plt.plot(x,l)
```

```python
data.boxplot(column="tip",by="sex")
plt.suptitle("")
plt.title("The box with 2 features for the tip")
plt.xlabel("Features of the sex")
plt.ylabel("Vaules of tip")
```

```python
data.boxplot(column="tip",by="smoker")
plt.suptitle("")
plt.title("The box with 2 features for the tip")
plt.xlabel("Features of the smoker")
plt.ylabel("Vaules of tip")
```

```python
data.boxplot(column="tip",by="day")
plt.suptitle("")
plt.title("The box with 4 features for the tip")
plt.xlabel("Features of the day")
plt.ylabel("Vaules of tip")
```

```python
data.boxplot(column="tip",by="time")
plt.suptitle("")
plt.title("The box with 2 features for the tip")
plt.xlabel("Features of the time")
plt.ylabel("Vaules of tip")
```

```python
data.boxplot(column="tip",by="size")
plt.suptitle("")
plt.title("The box with 6 features for the tip")
plt.xlabel("Features of the size")
plt.ylabel("Vaules of tip")
```

```python
data.groupby(["sex","smoker"],observed=False)["tip"].mean().unstack().plot(kind="bar")
plt.title('Values by Sex (Male or Female) and Smoker (Yes or No)')
plt.xlabel("Sex")
plt.ylabel("Value of tip")
```
