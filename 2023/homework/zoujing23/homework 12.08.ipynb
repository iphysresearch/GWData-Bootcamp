```python
# 包装一个散点图的函数便于复用
def scatterplot1(x_data, y_data, x_label, y_label, title, ax = None):

    # 创建一个绘图对象
    if ax:
        pass
    else:
        fig, ax = plt.subplots()
    # 调用plt句柄画一个图，返回两个变量fig，ax，分别是figure和 axes的缩写。函数返回一个figure图像和一个子图ax的array列表。
    # 一个fig图里边会套好几个ax，而每个ax会负责当前坐标上的图。所以任何绘制的图形都是在ax上完成的；
    # 而设置整个画布，将是在fig上完成。
    
    # 不显示顶部和右侧的坐标线
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    # 设置数据x_data和y_data、点的大小s、点的颜色color和透明度alpha
    ax.scatter(x_data, y_data, s = 10, color = '#539caf', alpha = 0.75)

    # 添加标题和坐标说明
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
#the first question
scatterplot1(x_data=iris.sepal_length,
            y_data=iris.petal_length,
           x_label='sepal_length',
           y_label='petal_length',
           title='sl vs pl')

scatterplot1(x_data=iris.sepal_width,
            y_data=iris.petal_width,
           x_label='sepal_width',
           y_label='petal_width',
           title='sw vs pw')

scatterplot1(x_data=iris.sepal_width,
            y_data=iris.petal_length,
           x_label='sepal_width',
           y_label='petal_length',
           title='sw vs pl')
scatterplot1(x_data=iris.sepal_length,
            y_data=iris.petal_width,
           x_label='sepal_length',
           y_label='petal_width',
           title='sl vs pw')
```

```python
# 包装一个散点图的函数便于复用
def scatterplot2(x_data1, y_data1, 
                x_data2, y_data2, 
                x_data3, y_data3, 
                x_label, y_label, title, ax = None):

    # 创建一个绘图对象
    if ax:
        pass
    else:
        fig, ax = plt.subplots()
    # 调用plt句柄画一个图，返回两个变量fig，ax，分别是figure和 axes的缩写。函数返回一个figure图像和一个子图ax的array列表。
    # 一个fig图里边会套几个ax，而每个ax会负责当前坐标上的图。所以任何绘制的图形都是在ax上完成的；
    # 而设置整个画布，将是在fig上完成。
    
    # 不显示顶部和右侧的坐标线
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    # 设置数据x_data和y_data、点的大小s、点的颜色color和透明度alpha
    ax.scatter(x_data1, y_data1, s = 10, color = 'b', alpha = 0.75)
    ax.scatter(x_data2, y_data2, s = 10, color = 'r', alpha = 0.75)
    ax.scatter(x_data3, y_data3, s = 10, color = 'g', alpha = 0.75)


    # 添加标题和坐标说明
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
def boxplot(x_data, y_data, base_color, median_color, x_label, y_label, title):
    _, ax = plt.subplots()

    # 设置样式
    ax.boxplot(y_data
               # 箱子是否颜色填充
               , patch_artist = True
               # 中位数线颜色
               , medianprops = {'color': base_color}
               # 箱子颜色设置，color：边框颜色，facecolor：填充颜色
               , boxprops = {'color': base_color, 'facecolor': median_color}
               # 猫须颜色whisker
               , whiskerprops = {'color': median_color}
               # 猫须界限颜色whisker cap
               , capprops = {'color': base_color})

    # 箱图与x_data保持一致
    ax.set_xticklabels(x_data)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
```

```python
#定义绘制分组柱状图的函数
def groupedbarplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 设置每一组柱状图的宽度
    total_width = 0.8
    # 设置每一个柱状图的宽度
    ind_width = total_width / len(y_data_list)
    # 计算每一个柱状图的中心偏移
    alteration = np.arange(-total_width/2+ind_width/2, total_width/2+ind_width/2, ind_width)

    # 分别绘制每一个柱状图
    for i in range(0, len(y_data_list)):
        # 横向散开绘制
        ax.bar(x_data + alteration[i], y_data_list[i], color = colors[i], label = y_data_names[i], width = ind_width)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right')
```

```python
#定义绘制堆积柱状图的函数
def stackedbarplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 循环绘制堆积柱状图
    for i in range(0, len(y_data_list)):
        if i == 0:
            ax.bar(x_data, y_data_list[i], color = colors[i], align = 'center', label = y_data_names[i])
        else:
            # 采用堆积的方式，除了第一个分类，后面的分类都从前一个分类的柱状图接着画
            # 用归一化保证最终累积结果为1，下面bottom参数表示纵向从哪里开始画
            ax.bar(x_data, y_data_list[i], color = colors[i], bottom = y_data_list[i - 1], align = 'center', label = y_data_names[i])
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right') # 设定图例位置
```

```python
#the first question
#处理数据，统计不同仓位等级幸存和遇难的人数
survived_1=np.sum((titanic['survived']==1)&(titanic['pclass']==1))
survived_2=np.sum((titanic['survived']==1)&(titanic['pclass']==2))
survived_3=np.sum((titanic['survived']==1)&(titanic['pclass']==3))
unsurvived_1=np.sum((titanic['survived']==0)&(titanic['pclass']==1))
unsurvived_2=np.sum((titanic['survived']==0)&(titanic['pclass']==2))
unsurvived_3=np.sum((titanic['survived']==0)&(titanic['pclass']==3))
tabel_1=pd.DataFrame({
    'survived':[survived_1,survived_2,survived_3],
    'unsurvived':[unsurvived_1,unsurvived_2,unsurvived_3],
    'total':[survived_1+unsurvived_1,survived_2+unsurvived_2,survived_3+unsurvived_3],
    'survived_prop':[(survived_1/(survived_1+unsurvived_1)),(survived_2/(survived_2+unsurvived_2)),(survived_3/(survived_3+unsurvived_3))],
    'unsurvived_prop':[(unsurvived_1/(survived_1+unsurvived_1)),(unsurvived_2/(survived_2+unsurvived_2)),(unsurvived_3/(survived_3+unsurvived_3))]
                     })
tabel_1.index=[1,2,3]
print(tabel_1)

#绘图
stackedbarplot(x_data = tabel_1.index.values
               , y_data_list = [tabel_1['survived_prop'], tabel_1['unsurvived_prop']]
               , y_data_names = ['survived', 'unsurvived']
               , colors = ['#5C3317', '#DB9370']
               , x_label = 'pclass'
               , y_label = 'Proportion of survived/unsurvived'
               , title = 'proportion of survived/unsurvived by pclass(1,2,3)')
plt.xticks([1,2,3],['1','2','3'])
```

```python
#second question
#处理数据，统计出不同性别的幸存人数和总人数
survived_male=np.sum((titanic['sex']=='male')&(titanic['survived']==1))
survived_female=np.sum((titanic['sex']=='female')&(titanic['survived']==1))
unsurvived_male=np.sum((titanic['sex']=='male')&(titanic['survived']==0))
unsurvived_female=np.sum((titanic['sex']=='female')&(titanic['survived']==0))
tabel_2=pd.DataFrame({
    'surived':[survived_male,survived_female],
    'unsurvived':[unsurvived_male,unsurvived_female],
    'total':[survived_male+unsurvived_male,survived_female+unsurvived_female],
    'survived_prop':[(survived_male/(survived_male+unsurvived_male)),(survived_female/(survived_female+unsurvived_female))],
    'unsurvived_prop':[(unsurvived_male/(survived_male+unsurvived_male)),(unsurvived_female/(survived_female+unsurvived_female))]
                    })

print(tabel_2)

#绘图
stackedbarplot(x_data =  tabel_2.index.values
               , y_data_list = [tabel_2['survived_prop'], tabel_2['unsurvived_prop']]
               , y_data_names = ['survived', 'unsurvived']
               , colors = ['#5C3317', '#DB9370']
               , x_label = 'sex'
               , y_label = 'Proportion of survived/unsurvived'
               , title = 'proportion of survived/unsurvived by sex(male,female)')
plt.xticks([0,1,],['male','female'])
```

```python
#the forth question
#处理数据，分别统计幸存和遇难乘客的年龄
survived_age=titanic[titanic['survived']==1].age
unsurvived_age=titanic[titanic['survived']==0].age
#观察发现数据中有空白值，将其去除
survived_age=survived_age.dropna()
unsurvived_age=unsurvived_age.dropna()
tabel_4=[unsurvived_age,survived_age]

#绘图
boxplot(x_data = [0,1]
        , y_data = tabel_4
        , base_color = 'y'
        , median_color = 'r'
        , x_label = 'features of survived'
        , y_label = 'values of age'
        , title = 'the box with 2 features for the age')
```

```python
#the fifth question
#处理数据，将不同上船港口与其乘客仓位等级统计出来
C_1=np.sum((titanic['pclass']==1)&(titanic['embarked']=='C'))
C_2=np.sum((titanic['pclass']==2)&(titanic['embarked']=='C'))
C_3=np.sum((titanic['pclass']==3)&(titanic['embarked']=='C'))
Q_1=np.sum((titanic['pclass']==1)&(titanic['embarked']=='Q'))
Q_2=np.sum((titanic['pclass']==2)&(titanic['embarked']=='Q'))
Q_3=np.sum((titanic['pclass']==3)&(titanic['embarked']=='Q'))
S_1=np.sum((titanic['pclass']==1)&(titanic['embarked']=='S'))
S_2=np.sum((titanic['pclass']==2)&(titanic['embarked']=='S'))
S_3=np.sum((titanic['pclass']==3)&(titanic['embarked']=='S'))
tabel_5=pd.DataFrame({
    '1':[C_1,Q_1,S_1],
    '2':[C_2,Q_2,S_2],
    '3':[C_3,Q_3,S_3]
                    })
tabel_5.index=['C','Q','S']
print(tabel_5)

#绘图
groupedbarplot(x_data =[1,2,3]
               , y_data_list = [tabel_5['1'],tabel_5['2'],tabel_5['3']]
               , y_data_names = ['C', 'Q','S']
               , colors = ['#5C3317', '#6B4226','#DB9370']
               , x_label = 'embarked'
               , y_label = 'number of people for different kind of pclass'
               , title = 'number of people for different kind of pclass by embarked')
plt.xticks([1,2,3],['1','2','3'])
```

```python
#the sixth question
#处理数据，分别统计幸存和遇难乘客的堂兄弟姐妹的数量
survived_sibsp=titanic[titanic['survived']==1].sibsp
unsurvived_sibsp=titanic[titanic['survived']==0].sibsp
#去除空白值
survived_sibsp=survived_sibsp.dropna()
unsurvived_sibsp=unsurvived_sibsp.dropna()
tabel_6=[unsurvived_sibsp,survived_sibsp]

#绘图
boxplot(x_data = [0,1]
        , y_data = tabel_6
        , base_color = 'y'
        , median_color = 'r'
        , x_label = 'features of survived'
        , y_label = 'values of sibsp'
        , title = 'the box with 2 features for the sibsp')
```

```python
#the eigth question
#处理数据，将单独乘船与不是单独乘船的幸存和死亡数量统计出来
survived_alone     =np.sum((titanic['survived']==1) & (titanic['alone']==True))
survived_notalone  =np.sum((titanic['survived']==1) & (titanic['alone']==False))
unsurvived_alone   =np.sum((titanic['survived']==0) & (titanic['alone']==True))
unsurvived_notalone=np.sum((titanic['survived']==0) & (titanic['alone']==False))

tabel_8 = pd.DataFrame({
    'survived':[survived_notalone,survived_alone],
    'unsurvived':[unsurvived_notalone,unsurvived_alone]
                    })
tabel_8.index=['not alone','alone']
print(tabel_8)

#绘图
groupedbarplot(x_data =[0,1]
               , y_data_list = [tabel_8['unsurvived'],tabel_8['survived']]
               , y_data_names = ['unsurvived', 'survived']
               , colors = ['#6B4226','#DB9370']
               , x_label = 'alone'
               , y_label = 'number of people '
               , title = 'number of people by whether be survived  alone or not')
plt.xticks([0,1],['not alone','alone'])
```
