## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
sns.lineplot(x='year', y='passengers',data=data)
plt.title('The number of passenger in each year')
plt.show()
```

```python
sns.barplot(x='month', y='passengers',data=data,errorbar=None)
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
sns.pairplot(data,x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
```

```python
fig, ax =plt.subplots(2,2,constrained_layout=True)
sns.scatterplot(x='sepal_length',y='petal_length',data=data,ax=ax[0][0],hue="species")
sns.scatterplot(x='sepal_length',y='petal_width',data=data,ax=ax[0][1],hue="species")
sns.scatterplot(x='sepal_width',y='petal_length',data=data,ax=ax[1][0],hue="species")
sns.scatterplot(x='sepal_width',y='petal_width',data=data,ax=ax[1][1],hue="species")
```

```python
sns.boxplot(data,x='species',y='sepal_length')
```

```python
sns.boxplot(data,x='species',y='sepal_width')
```

```python
sns.boxplot(data,x='species',y='petal_length')
```

```python
sns.boxplot(data,x='species',y='petal_width' )
```
