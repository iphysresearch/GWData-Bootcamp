## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

### 数据导入

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
data.info()
```

### 设置全局参数

```python
import matplotlib

# 设置图片尺寸
matplotlib.rc('figure', figsize = (8, 6)) 

# 设置字体
matplotlib.rc('font', size = 14, family = 'Times')

# 不显示网格
matplotlib.rc('axes', grid = False) 

# 设置背景
matplotlib.rc('axes', facecolor = 'white') 
```

### 题目解答与绘图

#### 1. 分析年度乘客总量的变化情况（提示：折线图）

```python
# 按年份对乘客数量求和
yearly_passengers = data.groupby('year')['passengers'].sum()

# 绘制折线图
plt.plot(yearly_passengers.index, yearly_passengers.values, color = "black", lw = 2)

# 添加标题和标签
plt.title("The number of passengers in each year")
plt.xlabel("year")
plt.ylabel("number")

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 年度乘客总量几乎呈线性增长趋势，说明民航行业在稳步发展中。
</span>

#### 2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
# 按月份对乘客数量求和
monthly_passengers = data.groupby('month')['passengers'].sum()

# 绘制折线图
plt.bar(monthly_passengers.index, monthly_passengers.values, color = "#8dd3c7")

# 添加标题和标签
plt.title("The number of passengers in each month")
plt.xlabel("month", fontsize=14)
plt.ylabel("number", fontsize=14)

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 月度乘客量最多的是夏季，尤其是七、八月份。
</span>

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

### 数据导入

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

```python
data.info()
```

### 设置全局参数

```python
import matplotlib

# 设置图片尺寸
matplotlib.rc('figure', figsize = (8, 8)) 

# 设置字体
matplotlib.rc('font', size = 10, family = 'Times')

# 不显示网格
matplotlib.rc('axes', grid = False) 

# 设置背景
matplotlib.rc('axes', facecolor = 'white') 
```

### 题目解答与绘图

#### 1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）

```python
# 创建一个2x2的图表布局
fig, axs = plt.subplots(2, 2) 

# sepal_length vs. petal_length
axs[0, 0].scatter(data['sepal_length'], data['petal_length'])
axs[0, 0].set_title('s_l VS. p_l')
axs[0, 0].set_xlabel('s_l')
axs[0, 0].set_ylabel('p_l')

# sepal_length vs. petal_width
axs[0, 1].scatter(data['sepal_length'], data['petal_width'])
axs[0, 1].set_title('s_l VS. p_w')
axs[0, 1].set_xlabel('s_l')
axs[0, 1].set_ylabel('p_w')

# sepal_width vs. petal_length
axs[1, 0].scatter(data['sepal_width'], data['petal_length'])
axs[1, 0].set_title('s_w VS. p_l')
axs[1, 0].set_xlabel('s_w')
axs[1, 0].set_ylabel('p_l')

# sepal_width vs. petal_width
axs[1, 1].scatter(data['sepal_width'], data['petal_width'])
axs[1, 1].set_title('s_w VS. p_w')
axs[1, 1].set_xlabel('s_w')
axs[1, 1].set_ylabel('p_w')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

```python
# 计算pearson相关系数
print("Pearson Correlation Coefficient between sepal_length and petal_length:", np.corrcoef(data['sepal_length'], data['petal_length'])[0, 1])
print("Pearson Correlation Coefficient between sepal_length and petal_width:", np.corrcoef(data['sepal_length'], data['petal_width'])[0, 1])
print("Pearson Correlation Coefficient between sepal_width and petal_length:", np.corrcoef(data['sepal_width'], data['petal_length'])[0, 1])
print("Pearson Correlation Coefficient between sepal_width and petal_width:", np.corrcoef(data['sepal_width'], data['petal_width'])[0, 1])
```

<span style="color:red;">
结论: sepal 的长度与 petal 的宽度和长度均呈正相关，且为极强相关；sepal 的宽度与 petal 的宽度和长度均呈负相关，且为中等程度相关和弱相关。
</span>

#### 2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）

```python
#查看有几类
data['species'].value_counts()
```

```python
# 创建一个2x2的图表布局
fig, axs = plt.subplots(2, 2) 

# sepal_length vs. petal_length
print("Pearson Correlation Coefficient between sepal_length and petal_length") # 提示下面打印出来的是Pearson系数
for species in data['species'].unique():       # data['species'].unique() 提取有几类
    subdata = data[data['species'] == species] # 然后提取每类的数据，循环画在一个图上
    print(species+":", np.corrcoef(subdata['sepal_length'], subdata['petal_length'])[0, 1]) # 打印各种类 Pearson 系数
    axs[0, 0].scatter(subdata['sepal_length'], subdata['petal_length'])
axs[0, 0].set_title('s_l VS. p_l')
axs[0, 0].set_xlabel('s_l')
axs[0, 0].set_ylabel('p_l')

# sepal_length vs. petal_width
print("Pearson Correlation Coefficient between sepal_length and petal_width")
for species in data['species'].unique():      
    subdata = data[data['species'] == species] 
    print(species+":", np.corrcoef(subdata['sepal_length'], subdata['petal_width'])[0, 1])
    axs[0, 1].scatter(subdata['sepal_length'], subdata['petal_width'])
axs[0, 1].set_title('s_l VS. p_w')
axs[0, 1].set_xlabel('s_l')
axs[0, 1].set_ylabel('p_w')

# sepal_width vs. petal_length
print("Pearson Correlation Coefficient between sepal_width and petal_length")
for species in data['species'].unique():      
    subdata = data[data['species'] == species]
    print(species+":", np.corrcoef(subdata['sepal_width'], subdata['petal_length'])[0, 1])
    axs[1, 0].scatter(subdata['sepal_width'], subdata['petal_length'])
axs[1, 0].set_title('s_w VS. p_l')
axs[1, 0].set_xlabel('s_w')
axs[1, 0].set_ylabel('p_l')

# sepal_width vs. petal_width
print("Pearson Correlation Coefficient between sepal_width and petal_width")
for species in data['species'].unique():      
    subdata = data[data['species'] == species]
    print(species+":", np.corrcoef(subdata['sepal_width'], subdata['petal_width'])[0, 1])
    axs[1, 1].scatter(subdata['sepal_width'], subdata['petal_width'])
axs[1, 1].set_title('s_w VS. p_w')
axs[1, 1].set_xlabel('s_w')
axs[1, 1].set_ylabel('p_w')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 
    
    1）对于 setosa, sepal 和 petal 的大小均为正相关，且均为弱相关或极弱相关。
    2）对于 versicolor, sepal 和 petal 的大小均为正相关，其中 sepal 的长度与 petal 长度和 sepal 的宽度与 petal 宽度均为强相关，其它为中等程度相关。
    3）对于 virginica, sepal 和 petal 的大小均为正相关，其中 sepal 的长度与 petal 的长度为极强相关，sepal 的长度与 petal 的宽度为弱相关，其余均为中等程度相关。
</span>

#### 3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
# 创建一个1x3的图表布局
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# setosa
num = 0
spec = 'setosa'
axs[num].boxplot([data[data['species'] == spec]['sepal_length'],
                   data[data['species'] == spec]['sepal_width'],
                   data[data['species'] == spec]['petal_length'],
                   data[data['species'] == spec]['petal_width']],
                   labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
axs[num].set_title('The box with 4 features for the '+ spec +' flowers')
axs[num].set_xlabel('Features of the '+ spec +' flowers')
axs[num].set_ylabel('Values')

# versicolor
num = 1
spec = 'versicolor'
axs[num].boxplot([data[data['species'] == spec]['sepal_length'],
                   data[data['species'] == spec]['sepal_width'],
                   data[data['species'] == spec]['petal_length'],
                   data[data['species'] == spec]['petal_width']],
                   labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
axs[num].set_title('The box with 4 features for the '+ spec +' flowers')
axs[num].set_xlabel('Features of the '+ spec +' flowers')
axs[num].set_ylabel('Values')

# virginica
num = 2
spec = 'virginica'
axs[num].boxplot([data[data['species'] == spec]['sepal_length'],
                   data[data['species'] == spec]['sepal_width'],
                   data[data['species'] == spec]['petal_length'],
                   data[data['species'] == spec]['petal_width']],
                   labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
axs[num].set_title('The box with 4 features for the '+ spec +' flowers')
axs[num].set_xlabel('Features of the '+ spec +' flowers')
axs[num].set_ylabel('Values')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 
    
    1）对于 sepal 的长度和宽度，三个种类的数据范围都在2左右，且宽度的中心值都在3附近，而长度排序为：virginica, versicolor, setosa.
    2）对于 petal 的长度, 大小关系为：virginica, versicolor, setosa; 集中程度为：setosa, versicolor, virginica; 且 setosa 中心值偏大，其它的比较对称。
    3）对于 petal 的宽度, 大小关系为：virginica, versicolor, setosa; 集中程度为：setosa, versicolor, virginica; 且它们的中心值均偏小。
</span>

### 题目解答与绘图

#### 1. 小费和总消费之间的关系（提示：散点图+回归分析）

```python
# total_bill vs. tip
plt.scatter(data['total_bill'],data['tip'])
plt.title('total_bill VS. tip')
plt.xlabel('total_bill')
plt.ylabel('tip')

# 计算 Pearson 系数
print("Pearson Correlation Coefficient between sepal_length and petal_length:", np.corrcoef(data['total_bill'], data['tip'])[0, 1])

# 计算回归线的系数
coefficients = np.polyfit(data['total_bill'], data['tip'], 1)
# 创建回归线的函数
polynomial = np.poly1d(coefficients)
# 计算对应于total_bill的tip预测值
line = polynomial(data['total_bill'].sort_values()) # .sort_values() 用于排序，因为 data['total_bill'] 不是有序的

# 绘制回归线
plt.plot(data['total_bill'].sort_values(), line, color='red')

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 总消费越高，给小费的倾向就更高，而且这种正相关是强相关。

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

### 数据导入

```python
data = sns.load_dataset("titanic")
data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

```python
print(len(data))
data.info()
```

### 设置全局参数

```python
import matplotlib

# 设置图片尺寸
matplotlib.rc('figure', figsize = (5, 5)) 

# 设置字体
matplotlib.rc('font', size = 10, family = 'Times')

# 不显示网格
matplotlib.rc('axes', grid = False) 

# 设置背景
matplotlib.rc('axes', facecolor = 'white') 
```

### 题目解答与绘图

#### 1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）

```python
# 添加遇难列
data['unsurvived'] = 1 - data['survived']  

# 按仓位等级分组
survied_unsurvied = data.groupby('pclass')[['survived', 'unsurvived']].sum()

# 计算生存率和遇难率
survied_unsurvied['total'] = survied_unsurvied['survived'] + survied_unsurvied['unsurvived']
survied_unsurvied['survived_prop'] = survied_unsurvied['survived'] / survied_unsurvied['total']
survied_unsurvied['unsurvived_prop'] = survied_unsurvied['unsurvived'] / survied_unsurvied['total']

# 打印结果
print(survied_unsurvied)
```

```python
# 绘制堆叠柱状图
fig, ax = plt.subplots()
survied_unsurvied[['survived_prop','unsurvived_prop']].plot(kind='bar', stacked=True, ax=ax, color=['#6f77b4', '#ff9f0e'])
    # kind='bar' 表示表绘制柱状图，
    # stacked=True 表示柱状图是堆叠的，
    # ax 是一个轴对象，用于添加标题、标签和其他自定义设置。

# 添加标题和标签
ax.set_title('Proportion of survived/unsurvived by Pclass (1, 2, 3)')
ax.set_xlabel('Pclass')
ax.set_ylabel('Proportion of survived/unsurvived')

# 将x轴的标签设置为水平
plt.xticks(rotation=0)

# 显示图形
plt.show()
```

<span style="color:red;">
结论: 仓位等级越高，生存率越高！
</span>

#### 2. 不同性别的幸存比例（提示：箱图或者提琴图）

```python
# 添加遇难列
data['unsurvived'] = 1 - data['survived']  

# 按性别分组
sex = data.groupby('sex')[['survived', 'unsurvived']].sum()

# 计算生存率和遇难率
sex['total'] = sex['survived'] + sex['unsurvived']
sex['survived_prop'] = sex['survived'] / sex['total']
sex['unsurvived_prop'] = sex['unsurvived'] / sex['total']

# 打印结果
print(sex)
```

```python
# 绘制堆叠柱状图
fig, ax = plt.subplots()
sex[['survived_prop','unsurvived_prop']].plot(kind='bar', stacked=True, ax=ax, color=['#6f77b4', '#ff9f0e'])
    # kind='bar' 表示表绘制柱状图，
    # stacked=True 表示柱状图是堆叠的，
    # ax 是一个轴对象，用于添加标题、标签和其他自定义设置。

# 添加标题和标签
ax.set_title('Proportion of survived/unsurvived by sex')
ax.set_xlabel('Sex')
ax.set_ylabel('Proportion of survived/unsurvived')

# 将x轴的标签设置为水平
plt.xticks(rotation=0)

# 显示图形
plt.show()
```

<span style="color:red;">
