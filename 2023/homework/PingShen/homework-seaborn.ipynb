## 练习1：航班乘客变化分析 (2个题)

1. 分析年度乘客总量的变化情况（提示：折线图）
2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
data1 = sns.load_dataset("flights")
#data1.head()
# 年份，月份，乘客数
```

### 1. 分析年度乘客总量的变化情况（提示：折线图）

```python
data1_year=data1.groupby('year')['passengers'].sum()
#data1_year
#type(data1_year)
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x="year", y="passengers", data=data1, errorbar=None) 
plt.title("the number of passengers in each year",fontsize=10)
plt.show()
```

### 2. 分析乘客量在一年中12个月份的分布（提示：柱状图）

```python
plt.figure(figsize=(6,4))
sns.barplot(x='month', y="passengers", data=data1, errorbar=None)
plt.title("the number of passengers in each month",fontsize=10) 
plt.show()
```

## 练习2：鸢尾花花型尺寸分析 (3个题)

1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）
2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）
3. 不同种类鸢尾花萼片和花瓣大小的分布情况（六角箱图或者核密度估计）

```python
data2 = sns.load_dataset("iris")
# data2.head()
# # 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
# data2.info()
```

```python
sns.pairplot(data2, hue='species')
```

### 1. 萼片（sepal）和花瓣（petal）的大小关系（提示：散点图）

```python
g = sns.PairGrid(data2 ,
                 x_vars=["sepal_length", "sepal_width"],
                 y_vars=["petal_length", "petal_length"],
                                  )
g.map(sns.scatterplot, palette="pastel");
```

### 2. 不同种类（species）鸢尾花萼片和花瓣的大小关系（提示：箱图或者提琴图）

```python
setosa=data2[data2.species=='setosa']
versicolor=data2[data2.species=='versicolor']
virginica=data2[data2.species=='virginica']
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=setosa.loc[:,["sepal_length", "sepal_width", "petal_length", "petal_width"]])
plt.title('features of setosa flowers')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=versicolor.loc[:,["sepal_length", "sepal_width", "petal_length", "petal_width"]])
plt.title('features of versicolor flowers')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=virginica.loc[:,["sepal_length", "sepal_width", "petal_length", "petal_width"]])
plt.title('features of virginica flowers')
plt.show()
```

### 3. 不同种类鸢尾花萼片和花瓣大小的分布情况

```python
g = sns.PairGrid(data2 ,
                 x_vars=["sepal_length", "sepal_width"],
                 y_vars=["petal_length", "petal_length"],
                 hue='species',
                 )
g.map(sns.scatterplot, palette="pastel")
```

## 练习4：泰坦尼克号海难幸存状况分析 (8个题)

1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）
2. 不同性别的幸存比例（提示：箱图或者提琴图）
3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）
4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）
5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）
6. 幸存和遇难乘客堂兄弟姐妹的数量分布（提示：箱图或者提琴图）
7. 幸存和遇难乘客父母子女的数量分布（提示：箱图或者提琴图）
8. 单独乘船与否和幸存之间有没有联系（提示：统计柱状图）

```python
data4 = sns.load_dataset("titanic")
data4.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

### 1. 不同仓位等级中幸存和遇难的乘客比例（提示：箱图或者提琴图）


```python
pclass_survived=data4.groupby(['pclass','survived']).size().unstack()

pclass_survived['total'] = pclass_survived[0] + pclass_survived[1]
#total_sum=pclass_survived['total'] .sum()
pclass_survived['0_percent'] = pclass_survived[0] / pclass_survived['total'] 
pclass_survived['1_percent'] = pclass_survived[1] / pclass_survived['total'] 
pclass_survived['total_percent'] = pclass_survived['total']  / pclass_survived['total'] 
pclass_survived
```

```python
plt.figure()
sns.barplot(x='pclass',y='total_percent',data=pclass_survived,label='unsurvived')
sns.barplot(x='pclass',y='1_percent',data=pclass_survived,label='survived')
plt.show()
```

### 2. 不同性别的幸存比例


```python
sex_survived=data4.groupby(['sex','survived']).size().unstack()
sex_survived['total'] = sex_survived[0] + sex_survived[1]
sex_survived['0_percent'] = sex_survived[0] / sex_survived['total'] 
sex_survived['1_percent'] = sex_survived[1] / sex_survived['total'] 
sex_survived['total_percent'] =sex_survived['total']  / sex_survived['total'] 
sex_survived
```

```python
plt.figure()
sns.barplot(x='sex',y='total_percent',data=sex_survived,label='survived')
sns.barplot(x='sex',y='1_percent',data=sex_survived,label='unsurvived')
plt.show()
```

### 3. 幸存和遇难乘客的票价分布（提示：箱图或者提琴图）


```python
sns.boxplot(x="survived", y="fare", data=data4)
```

### 4. 幸存和遇难乘客的年龄分布（提示：箱图或者提琴图）


```python
sns.boxplot(x="survived", y="age", data=data4)
```

### 5. 不同上船港口的乘客仓位等级分布（提示：箱图或者提琴图）


```python
data4['other']=np.ones(len(data4))
sns.barplot(x="pclass", y="other", hue="embarked", estimator="sum", data=data4)
```
